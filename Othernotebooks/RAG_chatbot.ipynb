````python
# ==============================
# DATA INGESTION PIPELINE
# ==============================

import json
import xml.etree.ElementTree as ET
from lxml import etree
from haystack import Document
from collections import Counter


# ==============================
# MITRE ATT&CK (Mobile + ICS)
# ==============================

def ingest_mitre(path, source):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    docs = []
    for obj in data.get("objects", []):
        if obj.get("type") == "attack-pattern":
            name = obj.get("name", "")
            desc = obj.get("description", "")

            if not desc:
                continue

            docs.append(
                Document(
                    content=f"MITRE Technique: {name}\nDescription: {desc}",
                    meta={
                        "source": source,
                        "stix_id": obj.get("id")
                    }
                )
            )
    return docs


mitre_docs = []
mitre_docs += ingest_mitre("mobile-attack (1).json", "MITRE_MOBILE")
mitre_docs += ingest_mitre("ics-attack (1).json", "MITRE_ICS")


# ==============================
# ATM
# ==============================

def ingest_atm(path):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    docs = []
    for obj in data.get("objects", []):
        if obj.get("type") == "attack-pattern":
            name = obj.get("name", "")
            desc = obj.get("description", "")

            if not desc:
                continue

            docs.append(
                Document(
                    content=f"ATM Technique: {name}\nDescription: {desc}",
                    meta={
                        "source": "ATM",
                        "stix_id": obj.get("id")
                    }
                )
            )
    return docs


atm_docs = ingest_atm("latest atm.json")


# ==============================
# CAPEC
# ==============================

def ingest_capec(xml_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    ns = {"capec": "http://capec.mitre.org/capec-3"}
    docs = []

    for ap in root.findall(".//capec:Attack_Pattern", ns):
        capec_id = ap.get("ID")
        name = ap.findtext("capec:Name", default="", namespaces=ns)
        desc = ap.findtext("capec:Description", default="", namespaces=ns)

        if not desc:
            continue

        docs.append(
            Document(
                content=f"CAPEC-{capec_id}: {name}\nDescription: {desc}",
                meta={
                    "source": "CAPEC",
                    "capec_id": capec_id
                }
            )
        )
    return docs


capec_docs = ingest_capec("capec_v3.9.xml")


# ==============================
# CWE
# ==============================

def ingest_cwe(xml_path):
    parser = etree.XMLParser(recover=True, huge_tree=True)
    tree = etree.parse(xml_path, parser)
    root = tree.getroot()

    ns = {"cwe": "http://cwe.mitre.org/cwe-7"}
    docs = []

    for w in root.findall(".//cwe:Weakness", namespaces=ns):
        cwe_id = w.get("ID")
        name = w.get("Name")

        desc_elem = w.find("cwe:Description", namespaces=ns)
        desc = desc_elem.text if desc_elem is not None else ""

        if not desc:
            continue

        docs.append(
            Document(
                content=f"CWE-{cwe_id}: {name}\nDescription: {desc}",
                meta={
                    "source": "CWE",
                    "cwe_id": f"CWE-{cwe_id}"
                }
            )
        )

    return docs


cwe_docs = ingest_cwe("cwec_v4.19.1.xml")


# ==============================
# MERGE ALL DOCUMENTS
# ==============================

all_docs = []
all_docs += mitre_docs
all_docs += atm_docs
all_docs += capec_docs
all_docs += cwe_docs


# ==============================
# SOURCE COUNT SUMMARY
# ==============================

print(Counter([d.meta["source"] for d in all_docs]))
````

````python
# ========== 6. LOAD & EMBED ALL DOCUMENTS (ISO + Annex + ECU + MITRE + ATM + CAPEC + CWE) ==========

import json, glob, os
from haystack import Document

# ------------------------------------------------------------------
# PATHS (adjust only if filenames change)
# ------------------------------------------------------------------
CLAUSE_PATH = "./clauses"          # clause-5.json, clause-6.json, ...
ANNEX_PATH  = "annex.json"         # Annex F
DATA_PATH   = "data.json"          # ECU / system data

# ------------------------------------------------------------------
# LOAD ECU / SYSTEM DATA
# ------------------------------------------------------------------
with open(DATA_PATH, "r", encoding="utf-8") as f:
    ecu_data = json.load(f)

ecu_docs = []
for k, v in ecu_data.items():
    ecu_docs.append(
        Document(
            content=json.dumps(v, indent=2),
            meta={
                "source": "ECU",
                "title": k
            }
        )
    )

print(f"âš™ï¸ Loaded {len(ecu_docs)} ECU documents")

# ------------------------------------------------------------------
# LOAD ISO 21434 CLAUSE JSONs
# ------------------------------------------------------------------
clause_files = sorted(glob.glob(os.path.join(CLAUSE_PATH, "clause-*.json")))
iso_docs = []

for file in clause_files:
    with open(file, "r", encoding="utf-8") as f:
        clause_content = json.load(f)

    clause_number = os.path.splitext(os.path.basename(file))[0].split("-")[1]

    iso_docs.append(
        Document(
            content=clause_content.get(
                "content",
                json.dumps(clause_content, indent=2)
            ),
            meta={
                "source": "ISO_21434",
                "clause": clause_number,
                "title": clause_content.get("title", f"Clause {clause_number}")
            }
        )
    )

print(f"ğŸ“˜ Loaded {len(iso_docs)} ISO 21434 clauses")

# ------------------------------------------------------------------
# LOAD ANNEX F
# ------------------------------------------------------------------
annex_docs = []

if os.path.exists(ANNEX_PATH):
    with open(ANNEX_PATH, "r", encoding="utf-8") as f:
        annex_json = json.load(f)

    content_parts = [
        f"Annex {annex_json.get('annex_id', '')}: {annex_json.get('annex_title', '')}"
    ]

    for section in annex_json.get("sections", []):
        content_parts.append(
            f"\n\n### {section.get('section_id')} â€” {section.get('section_title')}"
        )

        for c in section.get("content", []):
            content_parts.append(f"- {c}")

        for t in section.get("tables", []):
            content_parts.append(
                f"\nTable {t.get('table_id')}: {t.get('table_title')}"
            )
            cols = t.get("columns", [])
            rows = t.get("rows", [])

            if cols and rows:
                content_parts.append(" | ".join(cols))
                content_parts.append("-" * 40)
                for r in rows:
                    content_parts.append(
                        " | ".join(str(r.get(col, "")) for col in cols)
                    )

        for note in section.get("notes", []):
            content_parts.append(f"Note: {note}")

    annex_docs.append(
        Document(
            content="\n".join(content_parts),
            meta={
                "source": "ANNEX_F",
                "annex": annex_json.get("annex_id"),
                "title": annex_json.get("annex_title")
            }
        )
    )

    print(f"ğŸ“— Loaded Annex F")

else:
    print("âš ï¸ Annex F not found â€“ skipping")

# ------------------------------------------------------------------
# MERGE ALL SOURCES
# (These lists MUST already exist from earlier cells)
# ------------------------------------------------------------------
# mitre_docs   -> MITRE ATT&CK (ICS + Mobile)
# atm_docs     -> Automotive Threat Matrix
# capec_docs   -> CAPEC
# cwe_docs     -> CWE

all_docs = []
all_docs += ecu_docs
all_docs += iso_docs
all_docs += annex_docs
all_docs += mitre_docs
all_docs += atm_docs
all_docs += capec_docs
all_docs += cwe_docs

print(f"ğŸ“„ Total documents before embedding: {len(all_docs)}")

# ------------------------------------------------------------------
# EMBED + STORE (ONCE â€” CORRECT WAY)
# ------------------------------------------------------------------
result = doc_embedder.run(documents=all_docs)
embedded_docs = result["documents"]

document_store.write_documents(embedded_docs)

print(f"âœ… {len(embedded_docs)} documents embedded and stored")

# ------------------------------------------------------------------
# SUMMARY BY SOURCE (VERY IMPORTANT CHECK)
# ------------------------------------------------------------------
from collections import Counter
summary = Counter(d.meta.get("source", "Unknown") for d in embedded_docs)
print("ğŸ“Š Document distribution:")
for k, v in summary.items():
    print(f"  {k}: {v}")
````

````python
%%writefile tara_backend.py
# ==============================================
# TARA BACKEND â€” STREAMLIT / PRODUCTION VERSION
# ISO + ECU + ANNEX + MITRE + ATM + CAPEC + CWE
# ==============================================

# ========== 1. IMPORTS ==========
import os
import json
import glob
from pathlib import Path

from haystack import Pipeline, Document
from haystack.document_stores.in_memory import InMemoryDocumentStore
from haystack.components.embedders import (
    SentenceTransformersDocumentEmbedder,
    SentenceTransformersTextEmbedder,
)
from haystack.components.retrievers.in_memory import InMemoryEmbeddingRetriever
from haystack.components.builders import PromptBuilder
from haystack_integrations.components.generators.google_ai import GoogleAIGeminiGenerator

print("âœ… Imports complete.")

# ========== 2. DOCUMENT STORE & EMBEDDERS ==========
document_store = InMemoryDocumentStore()
print("ğŸ§  Document store initialized.")

doc_embedder = SentenceTransformersDocumentEmbedder(
    model="sentence-transformers/all-MiniLM-L6-v2"
)
text_embedder = SentenceTransformersTextEmbedder(
    model="sentence-transformers/all-MiniLM-L6-v2"
)

doc_embedder.warm_up()
print("âœ… Embedders ready.")

retriever = InMemoryEmbeddingRetriever(document_store)
print("ğŸ” Retriever initialized.")

# ========== 3. STRICT ISO PROMPT ==========
# ========== 3. STRICT ISO-LOCKED TEMPLATE ==========
template = """
You are an **automotive cybersecurity analyst** performing a **Threat Analysis and Risk Assessment (TARA)**
strictly according to **ISO/SAE 21434 Clause 15**.

If a property is NOT applicable, explicitly mark it as **Not Applicable (N/A)**.
DO NOT omit any property.

Your task is to generate a **professional, deterministic, and ISO-compliant TARA report**
in **Markdown format ONLY** for the given ECU or system.

âš ï¸ STRICT RULES (DO NOT VIOLATE):
- Follow the structure EXACTLY as provided
- Do NOT rename sections
- Do NOT add or remove sections
- Do NOT change definitions
- Do NOT invent new scoring models
- Do NOT output explanations, disclaimers, or commentary
- Do NOT vary terminology across runs

Failure to follow these rules makes the output INVALID.

---

### âš™ï¸ Rules for Reasoning (MANDATORY INTELLIGENCE MAPPING)

For EVERY threat, you MUST use:

- **CWE** â†’ Identify the underlying **technical root cause weakness**
- **CAPEC** â†’ Describe the **attack mechanism** used by the attacker
- **MITRE ATT&CK** â†’ Define the **attack path and entry point**
- **Automotive Threat Matrix (ATM)** â†’ Ensure the threat is **vehicle-relevant**
- **ISO/SAE 21434** â†’ Perform **risk scoring and compliance evaluation**

Threat logic MUST follow this chain:

**CWE (weakness) â†’ CAPEC (how exploited) â†’ MITRE (where/how attacker enters) â†’ ATM (why relevant to vehicle) â†’ ISO 21434 (risk & treatment)**

Do NOT skip any of these elements.

---

## âœ… ISO DEFINITIONS (USE EXACTLY â€” DO NOT CHANGE)

### Impact Categories (1 = Low â†’ 5 = Severe)
- Safety
- Operational
- Financial
- Privacy
- Regulatory

### Attack Feasibility Factors (1â€“5 each)
- Required Expertise
- Required Equipment
- Required Knowledge
- Required Access
- Required Time

### Calculations
- **Feasibility Score = (sum of factors Ã· 5)** (rounded to nearest integer)
- **Risk Score = Impact Ã— Feasibility**
- **Risk Classification**
  - 1â€“5 â†’ Low
  - 6â€“10 â†’ Medium
  - 11â€“25 â†’ High

---

## âœ… Cybersecurity Properties (MANDATORY â€” USE ALL)

For EVERY asset and EVERY threat, evaluate ALL properties below.
If not applicable, explicitly mark **N/A**.

1. Confidentiality
2. Integrity
3. Availability
4. Authenticity
5. Authorization
6. Accountability
7. Non-Repudiation

---

## ğŸ“˜ ISO CONTEXT (AUTHORITATIVE â€” DO NOT OVERRIDE)
{% for document in documents %}
- {{ document.content }}
{% endfor %}

System / ECU under analysis: **{{question}}**

---

# âœ… TARA REPORT â€” {{question}}

---

## 1. Item Definition (Clause 9)
- Purpose / Function
- Operational Environment
- Interfaces
- Dependencies

---

## 2. Asset Identification (Clause 15.3.2)

For EACH asset, produce EXACTLY **7 rows** (one per cybersecurity property).

| Asset | Cybersecurity Property | Status (Impacted / N/A) | Rationale |
|------|------------------------|-------------------------|-----------|

---

## 3. Threat and Risk Assessment (Clauses 15.3.3 â€“ 15.3.7)

Generate **AT LEAST 5 distinct threats**, each with a different attack vector.

---

### Threat ID: TS_<ECU>_00X â€” <Threat Title>

| Attribute | Value |
|----------|-------|
| Entry Point (MITRE) | Attack vector & entry path |
| Attack Mechanism (CAPEC) | How the attack is executed |
| Root Cause Weakness (CWE) | Underlying exploitable weakness |
| Vehicle Relevance (ATM) | Why this threat is valid in automotive context |
| Associated Asset | Asset |

---

### Cybersecurity Properties Violated
- Confidentiality: Yes / No / N/A
- Integrity: Yes / No / N/A
- Availability: Yes / No / N/A
- Authenticity: Yes / No / N/A
- Authorization: Yes / No / N/A
- Accountability: Yes / No / N/A
- Non-Repudiation: Yes / No / N/A

---

### Damage Scenarios
1. Safety Impact
2. Operational Impact
3. Privacy Impact
4. Financial Impact
5. Regulatory Impact

---

### Impact Rating (1â€“5)
Provide ONE value based on the most severe damage scenario.

---

### Attack Feasibility Factors (1â€“5 each)

| Factor | Rating |
|-------|--------|
| Required Expertise | # |
| Required Equipment | # |
| Required Knowledge | # |
| Required Access | # |
| Required Time | # |

â–º **Feasibility Score = (sum Ã· 5)**
â–º **Risk Score = Impact Ã— Feasibility**
â–º **Risk Classification = Low / Medium / High**

---

### Mitigation / Cybersecurity Goals (Clauses 10 & 11)
Map controls explicitly to violated cybersecurity properties AND ensure mitigations address:
- The **CWE weakness**
- The **CAPEC attack method**
- The **MITRE attack path**

---

## 4. Residual Risk (Clauses 8 & 11)
Summarize remaining risks and required verification activities.

---
"""

prompt_builder = PromptBuilder(
    template=template,
    required_variables=["documents", "question"]
)

print("ğŸ§¾ PromptBuilder initialized (STRICT ISO MODE).")

# ========== 4. GEMINI ==========
if "GOOGLE_API_KEY" not in os.environ:
    raise EnvironmentError("âŒ GOOGLE_API_KEY not set")

generator = GoogleAIGeminiGenerator(
    model="gemini-2.5-flash-lite"
)

print("ğŸ¤– Gemini initialized.")

# ========== 5. PIPELINE ==========
basic_rag_pipeline = Pipeline()

basic_rag_pipeline.add_component("text_embedder", text_embedder)
basic_rag_pipeline.add_component("retriever", retriever)
basic_rag_pipeline.add_component("prompt_builder", prompt_builder)
basic_rag_pipeline.add_component("llm", generator)

basic_rag_pipeline.connect("text_embedder.embedding", "retriever.query_embedding")
basic_rag_pipeline.connect("retriever", "prompt_builder.documents")
basic_rag_pipeline.connect("prompt_builder", "llm")

print("âœ… RAG pipeline built.")

# ========== 6. LOAD DOCUMENTS ==========
CLAUSE_PATH = "./clauses"
ANNEX_PATH = "annex.json"
DATA_PATH = "data.json"

docs = []

# ---------- ECU ----------
with open(DATA_PATH, "r", encoding="utf-8") as f:
    ecu_data = json.load(f)

for name, data in ecu_data.items():
    docs.append(
        Document(
            content=json.dumps(data, indent=2),
            meta={"source": "ECU", "title": name}
        )
    )

print(f"âš™ï¸ ECU docs loaded: {len(ecu_data)}")

# ---------- ISO CLAUSES ----------
for file in glob.glob(os.path.join(CLAUSE_PATH, "clause-*.json")):
    with open(file, "r", encoding="utf-8") as f:
        clause = json.load(f)

    clause_id = Path(file).stem
    docs.append(
        Document(
            content=clause.get("content", json.dumps(clause, indent=2)),
            meta={"source": "ISO_21434", "clause": clause_id}
        )
    )

print("ğŸ“˜ ISO clauses loaded.")

# ---------- ANNEX ----------
if os.path.exists(ANNEX_PATH):
    with open(ANNEX_PATH, "r", encoding="utf-8") as f:
        annex = json.load(f)

    docs.append(
        Document(
            content=json.dumps(annex, indent=2),
            meta={"source": "ANNEX_F"}
        )
    )
    print("ğŸ“— Annex F loaded.")

# ---------- HELPER: LOAD PREBUILT THREAT DOCS ----------
def load_docs(path, source):
    loaded = []
    if not os.path.exists(path):
        print(f"âš ï¸ Missing {path}")
        return loaded

    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    for d in data:
        loaded.append(
            Document(
                content=d["content"],
                meta={**d.get("meta", {}), "source": source}
            )
        )
    return loaded

# ---------- THREAT INTEL ----------
docs += load_docs("mitre_docs.json", "MITRE")
docs += load_docs("atm_docs.json", "ATM")
docs += load_docs("capec_docs.json", "CAPEC")
docs += load_docs("cwe_docs.json", "CWE")

print(f"ğŸ“„ Total documents before embedding: {len(docs)}")

# ========== 7. EMBED & STORE (ONCE) ==========
embedded_docs = doc_embedder.run(documents=docs)["documents"]
document_store.write_documents(embedded_docs)

print(f" {len(embedded_docs)} documents embedded & stored.")

# ========== 8. EXPORT ==========
__all__ = ["basic_rag_pipeline"]
````

Internal verification block

````python
%%writefile tara_verification.py
from typing import Dict, List

REQUIRED_PROPERTIES = [
    "Confidentiality",
    "Integrity",
    "Availability",
    "Authenticity",
    "Authorization",
    "Accountability",
    "Non-Repudiation",
]

def _has_retrieval_hit(retriever, query: str, keywords: List[str]) -> bool:
    """
    Query vector DB correctly (Haystack 2.x):
    TEXT â†’ EMBEDDING â†’ RETRIEVAL â†’ KEYWORD CHECK
    """

    # 1ï¸âƒ£ Get embedder attached to retriever
    embedder = getattr(retriever, "embedder", None)
    if embedder is None:
        return False

    # 2ï¸âƒ£ Convert text query â†’ embedding
    emb_result = embedder.run(text=query)
    query_embedding = emb_result["embedding"]

    # 3ï¸âƒ£ Retrieve documents using embedding
    docs = retriever.run(
        query_embedding=query_embedding,
        top_k=5
    ).get("documents", [])

    combined_text = " ".join(d.content.lower() for d in docs)

    # 4ï¸âƒ£ Keyword grounding check
    return any(k.lower() in combined_text for k in keywords)


def verify_threat_chain(threat: Dict, retriever) -> Dict:
    """
    Verify one threat against:
    CWE â†’ CAPEC â†’ MITRE â†’ ATM â†’ ISO 21434 logic
    """

    result = {
        "threat_id": threat.get("threat_id"),
        "structure_complete": True,
        "cwe_valid": False,
        "capec_valid": False,
        "mitre_valid": False,
        "atm_valid": False,
        "iso_consistent": True,
    }

    # ------------------ Structure ------------------
    mandatory_fields = [
        "entry_point",
        "associated_asset",
        "violated_properties",
        "impact_rating",
        "attack_feasibility_factors",
        "feasibility_score",
        "risk_score",
        "risk_classification",
    ]

    for f in mandatory_fields:
        if not threat.get(f):
            result["structure_complete"] = False

    # 7 cybersecurity properties
    for prop in REQUIRED_PROPERTIES:
        if prop not in threat.get("violated_properties", {}):
            result["structure_complete"] = False

    # ------------------ CWE ------------------
    result["cwe_valid"] = _has_retrieval_hit(
        retriever,
        query=threat.get("title", ""),
        keywords=["cwe-", "weakness", "improper", "unauthorized"]
    )

    # ------------------ CAPEC ------------------
    result["capec_valid"] = _has_retrieval_hit(
        retriever,
        query=threat.get("title", ""),
        keywords=["capec-", "attack pattern", "exploit", "bypass"]
    )

    # ------------------ MITRE ------------------
    result["mitre_valid"] = _has_retrieval_hit(
        retriever,
        query=threat.get("title", ""),
        keywords=["t08", "attack technique", "spoof", "modify"]
    )

    # ------------------ ATM ------------------
    result["atm_valid"] = _has_retrieval_hit(
        retriever,
        query=threat.get("associated_asset", ""),
        keywords=["ecu", "can", "lin", "flexray", "ota", "vehicle"]
    )

    # ------------------ ISO Risk Logic ------------------
    try:
        impact = int(threat["impact_rating"])
        feasibility = int(threat["feasibility_score"])
        computed_risk = impact * feasibility

        if str(computed_risk) != str(threat["risk_score"]):
            result["iso_consistent"] = False
    except Exception:
        result["iso_consistent"] = False

    return result


def verify_full_tara_report(tara_json: Dict, retriever) -> Dict:
    """
    Verify entire TARA report (internal use only)
    """

    verification = {
        "ecu": tara_json.get("ecu"),
        "threat_results": [],
        "overall_pass": True,
    }

    for threat in tara_json.get("threats", []):
        res = verify_threat_chain(threat, retriever)
        verification["threat_results"].append(res)

        if not all(res.values()):
            verification["overall_pass"] = False

    return verification
````

````python
%%writefile app.py
# app.py
import streamlit as st
from tara_backend import basic_rag_pipeline
from io import BytesIO
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Preformatted,
    PageBreak,
    Table,
    TableStyle,
)
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib import colors
from reportlab.lib.units import mm
from datetime import datetime
import markdown2
import json
import re
import html as html_lib
import unicodedata
from tara_verification import verify_full_tara_report

# ------------------------------
# ECU Dictionary (same as yours)
# ------------------------------
ECUS = {
    "abs": {"name": "ABS (Anti-lock Braking System)", "type": "ECU"},
    "bms": {"name": "BMS (Battery Management System)", "type": "ECU"},
    "tcu": {"name": "TCU (Telematics Control Unit)", "type": "ECU"},
    "telematics": {"name": "Telematics System", "type": "System"},
    "ecs": {"name": "ECU - General (Electronic Control Unit)", "type": "ECU"},
    "ecm": {"name": "ECM / Powertrain (Engine Control Module)", "type": "ECU"},
    "adas": {"name": "ADAS ECU", "type": "ECU"},
    "infotainment": {"name": "Infotainment Head Unit", "type": "ECU"},
    "gateway": {"name": "Gateway / Domain Controller", "type": "ECU"},
    "eps": {"name": "EPS (Electric Power Steering)", "type": "ECU"},
    "airbag": {"name": "Airbag / SRS ECU", "type": "ECU"},
    "bcm": {"name": "BCM (Body Control Module)", "type": "ECU"},
    "charger": {"name": "EV Charger Interface / Onboard Charger", "type": "Interface"},
    "motor_controller": {"name": "Motor Controller / Inverter", "type": "ECU"},
    "cluster": {"name": "Instrument Cluster", "type": "ECU"},
    "door_control": {"name": "Door Control Module", "type": "ECU"},
    "keyless_entry": {"name": "Keyless Entry / Smart Key System", "type": "Interface"},
    "hvac": {"name": "HVAC Controller", "type": "ECU"},
    "ota_server": {"name": "OTA Server / Update System", "type": "System"},
    "diagnostic": {"name": "Diagnostic Interface (OBD-II / JTAG / UART)", "type": "Interface"},
    "v2x": {"name": "V2X Communication Module", "type": "Interface"},
    "gps": {"name": "GNSS / GPS Receiver", "type": "Interface"},
    "lidar": {"name": "Lidar Sensor System", "type": "System"},
    "radar": {"name": "Radar Sensor System", "type": "System"},
    "camera": {"name": "Camera Sensor System", "type": "System"},
    "tpms": {"name": "TPMS (Tire Pressure Monitoring System)", "type": "System"},
    "obd": {"name": "OBD-II Diagnostic Port", "type": "Interface"},
    "bluetooth": {"name": "Bluetooth Interface", "type": "Interface"},
    "wifi": {"name": "Automotive Wi-Fi Module", "type": "Interface"},
    "can_bus": {"name": "CAN Bus Network", "type": "Interface"},
    "lin_bus": {"name": "LIN Bus Network", "type": "Interface"},
    "flexray": {"name": "FlexRay Network", "type": "Interface"},
    "zonal": {"name": "Zonal ECU", "type": "ECU"},
    "sensor_fusion": {"name": "Sensor Fusion Module", "type": "ECU"},
    "brake_control": {"name": "Brake Control System", "type": "System"},
    "throttle": {"name": "Throttle Control System", "type": "System"},
    "steering": {"name": "Steering ECU", "type": "ECU"},
    "charging_port": {"name": "Automotive Charging Port", "type": "Interface"},
    "mobile_app": {"name": "Vehicle Mobile App", "type": "System"},
    "voice_assistant": {"name": "Automotive Voice Assistant Integration", "type": "System"},
    "rfid_entry": {"name": "RFID Entry System", "type": "Interface"},
    "immobilizer": {"name": "Immobilizer System", "type": "Interface"},
    "ivi": {"name": "IVI (In-Vehicle Infotainment) System", "type": "System"},
    "digital_cockpit": {"name": "Digital Cockpit", "type": "System"},
    "usb": {"name": "USB Interface", "type": "Interface"},
    "hdmi": {"name": "HDMI Port (Automotive IVI)", "type": "Interface"},
    "media_interface": {"name": "Media Player Interface", "type": "System"},
    "e_call": {"name": "E-call System", "type": "System"},
    "telematics_cloud": {"name": "Telematics Cloud API", "type": "System"},
    "vehicle_config": {"name": "Vehicle Configuration API", "type": "System"},
}

# ------------------------------
# Streamlit UI Setup
# ------------------------------
st.set_page_config(page_title="TARA Report Generator", layout="wide")
st.title("ISO/SAE 21434 - TARA Report Generator")
st.markdown("Generate structured Threat Analysis and Risk Assessment (TARA) reports for ECUs using Haystack + Gemini RAG pipeline.")

# ECU Selection
ecu_display_names = [f"{v['name']} [{v['type']}]" for v in ECUS.values()]
ecu_keys = list(ECUS.keys())
selected_display = st.selectbox("Select ECU:", ecu_display_names)
selected_ecu = ECUS[ecu_keys[ecu_display_names.index(selected_display)]]["name"]

if "tara_report" not in st.session_state:
    st.session_state["tara_report"] = ""
    st.session_state["last_ecu"] = ""

# ------------------------------
# Generate TARA Report (unchanged)
# ------------------------------
generate_clicked = st.button("Generate Report", use_container_width=True)
if generate_clicked:
    with st.spinner("Generating report... please wait"):
        try:
            response = basic_rag_pipeline.run(
                {"text_embedder": {"text": selected_ecu}, "prompt_builder": {"question": selected_ecu}}
            )
            report = response["llm"]["replies"][0]
            st.session_state["tara_report"] = report
            st.session_state["last_ecu"] = selected_ecu


            st.success("TARA report generated successfully!")
        except Exception as e:
            st.error(f"Error generating report: {e}")

# ------------------------------
# Report Preview (Single Box Style, unchanged)
# ------------------------------
if st.session_state["tara_report"]:
    st.markdown("**Report Preview:**")

    st.markdown("""
    <style>
    .tara-report-box {
        background-color: #f9f9f9;
        border: 1px solid #ccc;
        border-radius: 10px;
        padding: 20px 25px;
        margin-top: 10px;
        margin-bottom: 25px;
        box-shadow: 0 1px 4px rgba(0,0,0,0.08);
        font-family: 'Segoe UI';
    }
    .tara-report-box h1, .tara-report-box h2, .tara-report-box h3 {
        color: #222;
        margin-top: 0;
    }
    .tara-report-box p {
        color: #333;
        font-size: 15px;
        line-height: 1.5;
    }
    </style>
    """, unsafe_allow_html=True)

    html_content = markdown2.markdown(
        st.session_state["tara_report"],
        extras=["fenced-code-blocks", "tables", "break-on-newline"]
    )
    html_content = re.sub(r"^<div[^>]*>|</div>\s*$", "", html_content.strip())

    st.markdown(f"<div class='tara-report-box'>{html_content}</div>", unsafe_allow_html=True)
else:
    st.info("No report generated yet. Choose an ECU and click Generate Report.")

# ------------------------------
# Helpers for PDF generation
# ------------------------------
def sanitize_filename(name: str):
    name = unicodedata.normalize("NFKD", name)
    name = re.sub(r"[^\w\s-]", "", name).strip()
    name = re.sub(r"[-\s]+", "_", name)
    return name

# Convert markdown to a list of ReportLab flowables; improved table handling
# Replace your existing markdown_to_story with this function
def markdown_to_story(md_text: str, styles, page_width, left_margin, right_margin):
    """
    Convert markdown to a ReportLab story (list of flowables).
    Improved table handling + HTML sanitization for table cells to avoid mismatched tags.
    """
    story = []

    # Convert Markdown -> HTML
    html = markdown2.markdown(md_text, extras=["fenced-code-blocks", "tables", "strike", "cuddled-lists"])
    html = re.sub(r"^<div[^>]*>", "", html)
    html = re.sub(r"</div>\s*$", "", html).strip()

    # Pattern to capture block-level elements
    pattern = re.compile(
        r'(?s)(<h[1-6].*?>.*?</h[1-6]>)|(<pre.*?>.*?</pre>)|(<table.*?>.*?</table>)|(<ul.*?>.*?</ul>)|(<ol.*?>.*?</ol>)|(<p.*?>.*?</p>)',
        flags=re.IGNORECASE,
    )
    matches = list(pattern.finditer(html))

    # fallback if nothing matched
    if not matches:
        cleaned = re.sub(r'<[^>]+>', '', html).strip()
        story.append(Paragraph(html_lib.unescape(cleaned), styles["Normal"]))
        return story

    def inline_replace(s: str) -> str:
        # convert common markdown-generated tags to ReportLab-friendly ones
        s = s.replace("<strong>", "<b>").replace("</strong>", "</b>")
        s = s.replace("<em>", "<i>").replace("</em>", "</i>")
        s = s.replace("<del>", "<u>").replace("</del>", "</u>")
        # convert inline code to font tag
        s = re.sub(r"<code>(.*?)</code>", r"<font name='Courier'>\1</font>", s, flags=re.DOTALL)
        s = s.replace("<br>", "<br/>")
        s = html_lib.unescape(s)
        return s

    def sanitize_and_balance_cell_html(c_html: str) -> str:
        """Sanitize HTML inside a table cell and balance allowed inline tags."""
        txt = c_html or ""
        txt = html_lib.unescape(txt)

        # Normalize some tags to simpler forms
        txt = txt.replace("<strong>", "<b>").replace("</strong>", "</b>")
        txt = txt.replace("<em>", "<i>").replace("</em>", "</i>")
        txt = txt.replace("<del>", "<u>").replace("</del>", "</u>")
        # handle code -> font
        txt = re.sub(r"<code[^>]*>(.*?)</code>", r"<font name='Courier'>\1</font>", txt, flags=re.DOTALL)

        # remove any attributes from b/i/u/br tags (not strictly necessary, just clean)
        txt = re.sub(r'<(\/?)(b|i|u|br)\b[^>]*>', r'<\1\2>', txt, flags=re.IGNORECASE)

        # Allow <font name='Courier'> ... </font> (we keep font tags as-is)
        # Remove all tags not in the allowed set {b,i,u,br,font}
        def tag_replacer(match):
            full = match.group(0)
            tagname = match.group(2).lower() if match.group(2) else ""
            # Keep allowed tags, pass through as-is (but strip unknown attrs except font)
            if tagname in ("b", "i", "u", "br"):
                return full  # already cleaned above to no attributes
            if tagname == "font":
                # keep font but strip unsafe attributes (keep name only)
                # extract name attribute if present
                name_match = re.search(r'name\s*=\s*["\"]([^"\"]+)["\"]', full, flags=re.IGNORECASE)
                if name_match:
                    return f"<font name='{name_match.group(1)}'>"
                # fallback: remove attributes
                return "<font>"
            # anything else -> remove tag entirely
            return ""
        txt = re.sub(r'<\s*(\/?)\s*([a-zA-Z0-9_:-]+)(?:\s+[^>]*)?>', tag_replacer, txt)

        # Now balance tags for allowed ones to avoid ReportLab parse errors
        def balance_tag(text: str, open_pattern: str, close_tag: str):
            open_count = len(re.findall(open_pattern, text, flags=re.IGNORECASE))
            close_count = text.lower().count(close_tag.lower())
            # remove extra closing tags
            while close_count > open_count:
                text = re.sub(re.escape(close_tag), "", text, count=1, flags=re.IGNORECASE)
                close_count -= 1
            # append missing closing tags
            while open_count > close_count:
                text = text + close_tag
                close_count += 1
            return text

        # Balance <b>, <i>, <u>, <font>
        txt = balance_tag(txt, r'<b(?:\s[^>]*)?>', '</b>')
        txt = balance_tag(txt, r'<i(?:\s[^>]*)?>', '</i>')
        txt = balance_tag(txt, r'<u(?:\s[^>]*)?>', '</u>')
        txt = balance_tag(txt, r'<font(?:\s[^>]*)?>', '</font>')

        # Finally, remove any leftover tags that somehow remain (safety)
        txt = re.sub(r'<(?![/]?[?]?(?:b|i|u|br|font)\b)[^>]+>', '', txt, flags=re.IGNORECASE)

        # Collapse excessive whitespace
        txt = re.sub(r'\s+', ' ', txt).strip()
        return txt

    # available width for tables
    usable_width = page_width - (left_margin + right_margin)

    for m in matches:
        block = m.group(0).strip()

        # Headings
        h = re.match(r'^<h([1-6])[^>]*>(.*?)</h\1>$', block, flags=re.IGNORECASE | re.DOTALL)
        if h:
            lvl = int(h.group(1))
            txt = html_lib.unescape(h.group(2).strip())
            txt = inline_replace(txt)
            if lvl == 1:
                story.append(Paragraph(txt, styles["Title"]))
            elif lvl == 2:
                story.append(Paragraph(txt, styles["Heading2"]))
            elif lvl == 3:
                h3 = ParagraphStyle(name="H3", parent=styles["Heading2"], fontSize=14, leading=16)
                story.append(Paragraph(txt, h3))
            else:
                hsmall = ParagraphStyle(name=f"H{lvl}", parent=styles["Normal"], fontSize=max(9, 14 - lvl))
                story.append(Paragraph(f"<b>{txt}</b>", hsmall))
            story.append(Spacer(1, 6))
            continue

        # Preformatted (code)
        if block.lower().startswith("<pre"):
            code_match = re.search(r'<pre[^>]*>(.*?)</pre>', block, flags=re.IGNORECASE | re.DOTALL)
            code_text = code_match.group(1) if code_match else block
            code_text = re.sub(r'^<code[^>]*>', '', code_text, flags=re.IGNORECASE)
            code_text = re.sub(r'</code>$', '', code_text, flags=re.IGNORECASE)
            code_text = html_lib.unescape(code_text)
            pre_style = ParagraphStyle(name="Code", fontName="Courier", fontSize=9, leading=12)
            story.append(Preformatted(code_text, pre_style))
            story.append(Spacer(1, 6))
            continue

        # Lists
        if block.lower().startswith("<ul") or block.lower().startswith("<ol"):
            items = re.findall(r'<li>(.*?)</li>', block, flags=re.IGNORECASE | re.DOTALL)
            ordered = block.lower().startswith("<ol")
            for i, it in enumerate(items, start=1):
                text = html_lib.unescape(it.strip())
                text = inline_replace(text)
                bullet = f"{i}." if ordered else "â€¢"
                item_style = ParagraphStyle(name="ListItem", parent=styles["Normal"], leftIndent=6)
                story.append(Paragraph(f"{bullet} {text}", item_style))
            story.append(Spacer(1, 6))
            continue

        # Tables: parse rows and render as ReportLab Table (with sanitized cells)
        if block.lower().startswith("<table"):
            rows = re.findall(r'<tr>(.*?)</tr>', block, flags=re.IGNORECASE | re.DOTALL)
            table_data = []
            max_cols = 0
            for row in rows:
                cells = re.findall(r'<t[dh][^>]*>(.*?)</t[dh]>', row, flags=re.IGNORECASE | re.DOTALL)
                clean_cells = []
                for c in cells:
                    # sanitize and balance cell HTML
                    c_text = sanitize_and_balance_cell_html(c)
                    # create Paragraph for the cell (allows inline bold/italic)
                    if c_text.strip() == "":
                        clean_cells.append(Paragraph("", styles["Normal"]))
                    else:
                        clean_cells.append(Paragraph(c_text, styles["Normal"]))
                if len(clean_cells) > max_cols:
                    max_cols = len(clean_cells)
                table_data.append(clean_cells)

            # normalize rows
            for r in table_data:
                if len(r) < max_cols:
                    r.extend([Paragraph("", styles["Normal"])] * (max_cols - len(r)))

            if max_cols == 0:
                continue

            col_width = usable_width / max_cols
            col_widths = [col_width] * max_cols

            tbl = Table(table_data, colWidths=col_widths, repeatRows=1)
            tbl_style = TableStyle([
                ("GRID", (0, 0), (-1, -1), 0.5, colors.HexColor("#cccccc")),
                ("BACKGROUND", (0, 0), (-1, 0), colors.HexColor("#f2f2f2")),
                ("VALIGN", (0, 0), (-1, -1), "TOP"),
                ("LEFTPADDING", (0, 0), (-1, -1), 6),
                ("RIGHTPADDING", (0, 0), (-1, -1), 6),
                ("TOPPADDING", (0, 0), (-1, -1), 4),
                ("BOTTOMPADDING", (0, 0), (-1, -1), 4),
            ])
            tbl.setStyle(tbl_style)
            story.append(tbl)
            story.append(Spacer(1, 8))
            continue

        # Paragraphs
        p_match = re.match(r'^<p[^>]*>(.*?)</p>$', block, flags=re.IGNORECASE | re.DOTALL)
        if p_match:
            inner = p_match.group(1).strip()
            inner = inline_replace(inner)
            story.append(Paragraph(inner, styles["Normal"]))
            story.append(Spacer(1, 6))
            continue

        # Fallback
        cleaned = re.sub(r'<[^>]+>', '', block).strip()
        if cleaned:
            story.append(Paragraph(html_lib.unescape(cleaned), styles["Normal"]))
            story.append(Spacer(1, 6))

    return story

def tara_markdown_to_structured_json(md_text: str, ecu_name: str):
    import re
    from datetime import datetime

    # ---------- ID REGEX ----------
    MITRE_RE = r"\bT\d{4}\b"
    CAPEC_RE = r"\bCAPEC-\d+\b"
    CWE_RE = r"\bCWE-\d+\b"
    ATM_RE = r"\bATM-\d+\b"

    def extract_ids(text, pattern):
        return list(set(re.findall(pattern, text)))

    def clean(v):
        return v.strip().strip("-").strip()

    def to_bool(v):
        v = v.lower()
        return v in ["yes", "true", "impacted"]

    def extract_number(text):
        m = re.findall(r"\d+", text)
        return m[0] if m else ""

    # ---------- BASE STRUCTURE ----------
    data = {
        "metadata": {
            "standard": "ISO/SAE 21434",
            "generated_at": datetime.now().isoformat()
        },
        "ecu": ecu_name,
        "item_definition": {
            "Purpose / Function": "",
            "Operational Environment": "",
            "Interfaces": "",
            "Dependencies": ""
        },
        "assets": [],
        "threats": [],
        "residual_risk": ""
    }

    lines = [l.rstrip() for l in md_text.splitlines() if l.strip()]
    i = 0
    section = None
    current_threat = None

    # ---------- PARSE ----------
    while i < len(lines):
        line = lines[i].strip()

        # ----- SECTION DETECTION -----
        if line.startswith("## 1. Item Definition"):
            section = "item"
            i += 1
            continue

        if line.startswith("## 2. Asset Identification"):
            section = "assets"
            i += 1
            continue

        if line.startswith("## 3. Threat and Risk Assessment"):
            section = "threats"
            i += 1
            continue

        if line.startswith("## 4. Residual Risk"):
            section = "residual"
            i += 1
            continue

        # ----- ITEM DEFINITION -----
        if section == "item" and ":" in line:
            k, v = line.split(":", 1)
            if k.strip() in data["item_definition"]:
                data["item_definition"][k.strip()] = v.strip()

        # ----- ASSETS -----
        if section == "assets" and line.startswith("|") and not line.startswith("| Asset"):
            cols = [c.strip() for c in line.split("|") if c.strip()]
            if len(cols) >= 4 and cols[0] != "------":
                data["assets"].append({
                    "asset": cols[0],
                    "security_property": cols[1],
                    "status": cols[2],
                    "rationale": cols[3]
                })

        # ----- THREAT START -----
        if line.startswith("### Threat ID:"):
            if current_threat:
                data["threats"].append(current_threat)

            tid_title = line.replace("### Threat ID:", "").strip()
            if "â€”" in tid_title:
                tid, title = tid_title.split("â€”", 1)
            else:
                tid, title = tid_title, ""

            current_threat = {
                "threat_id": tid.strip(),
                "title": title.strip(),
                "attributes": {
                    "entry_point_mitre": [],
                    "attack_mechanism_capec": [],
                    "root_cause_cwe": [],
                    "vehicle_relevance_atm": []
                },
                "associated_asset": "",
                "violated_properties": {},
                "impact_rating": "",
                "attack_feasibility_factors": {},
                "feasibility_score": "",
                "risk_score": "",
                "risk_classification": ""
            }

        # ----- THREAT DETAILS -----
        if section == "threats" and current_threat:

            # Attribute rows â†’ ID extraction
            if line.startswith("| Entry Point") or line.startswith("| Attack Mechanism") \
               or line.startswith("| Root Cause") or line.startswith("| Vehicle Relevance"):
                text = lines[i + 1]

                current_threat["attributes"]["entry_point_mitre"] += extract_ids(text, MITRE_RE)
                current_threat["attributes"]["attack_mechanism_capec"] += extract_ids(text, CAPEC_RE)
                current_threat["attributes"]["root_cause_cwe"] += extract_ids(text, CWE_RE)
                current_threat["attributes"]["vehicle_relevance_atm"] += extract_ids(text, ATM_RE)

            # Associated asset
            if line.startswith("| Associated Asset"):
                asset = clean(lines[i + 1])
                if asset and "|" not in asset:
                    current_threat["associated_asset"] = asset

            # CIAAN properties
            if ":" in line and any(p in line for p in [
                "Confidentiality", "Integrity", "Availability",
                "Authenticity", "Authorization",
                "Accountability", "Non-Repudiation"
            ]):
                k, v = line.split(":", 1)
                prop = k.replace("-", "").strip()
                current_threat["violated_properties"][prop] = to_bool(v.strip())

            # Impact rating
            if re.fullmatch(r"[1-5]", line):
                current_threat["impact_rating"] = line

            # Feasibility factors
            if line.startswith("| Required"):
                cols = [c.strip() for c in line.split("|") if c.strip()]
                if len(cols) == 2:
                    current_threat["attack_feasibility_factors"][cols[0]] = extract_number(cols[1])

            # Scores
            if "Feasibility Score" in line:
                current_threat["feasibility_score"] = extract_number(line)

            if "Risk Score" in line:
                current_threat["risk_score"] = extract_number(line)

            if "Risk Classification" in line:
                current_threat["risk_classification"] = line.split("=")[-1].strip().replace("*", "")

        # ----- RESIDUAL RISK -----
        if section == "residual":
            data["residual_risk"] += line + " "

        i += 1

    if current_threat:
        data["threats"].append(current_threat)

    data["residual_risk"] = data["residual_risk"].strip()

    return data


report_content = st.session_state.get("tara_report", "")

if report_content:
    tara_json = tara_markdown_to_structured_json(
        report_content,
        selected_ecu
    )

    try:
        # Haystack-official way (this is the key fix)
        retriever = basic_rag_pipeline.get_component("retriever")

        verification = verify_full_tara_report(
            tara_json=tara_json,
            retriever=retriever
        )

        # Internal only â€” NOT shown in UI, NOT exported
        with open("tara_verification_log.json", "w") as f:
            json.dump(verification, f, indent=2)

    except Exception as e:
        st.warning(f"Internal verification skipped: {e}")


# ------------------------------
# Export Buttons (improved PDF, landscape)
# ------------------------------
if st.session_state["tara_report"]:
    st.markdown("---")
    left, right = st.columns(2, gap="large")
    with left:
        if st.button("Export to PDF (improved formatting - landscape)", key="export_pdf_landscape"):
            # Landscape A4
            page_size = landscape(A4)
            page_width, page_height = page_size
            left_margin = right_margin = 20 * mm
            top_margin = bottom_margin = 20 * mm

            pdf_buffer = BytesIO()
            doc = SimpleDocTemplate(
                pdf_buffer,
                pagesize=page_size,
                rightMargin=right_margin,
                leftMargin=left_margin,
                topMargin=top_margin,
                bottomMargin=bottom_margin,
            )

            styles = getSampleStyleSheet()
            # Customize styles for headings and normal text
            styles["Title"].fontSize = 20
            styles["Title"].leading = 24
            styles["Heading2"].fontSize = 16
            styles["Heading2"].leading = 20
            styles["Normal"].fontSize = 10
            styles["Normal"].leading = 13

            # Build story
            story = []
            story.append(Paragraph("ISO/SAE 21434 - Threat Analysis and Risk Assessment (TARA) Report", styles["Title"]))
            story.append(Spacer(1, 6))
            story.append(Paragraph(f"<b>ECU:</b> {st.session_state.get('last_ecu', selected_ecu)}", styles["Heading2"]))
            story.append(Spacer(1, 8))

            md_text = st.session_state["tara_report"]
            story += markdown_to_story(md_text, styles, page_width, left_margin, right_margin)

            story.append(Spacer(1, 10))
            story.append(Paragraph(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}", styles["Normal"]))

            doc.build(story)
            pdf_buffer.seek(0)

            safe_name = sanitize_filename(st.session_state.get("last_ecu", selected_ecu))
            st.download_button("Download PDF", data=pdf_buffer, file_name=f"TARA_Report_{safe_name}_landscape.pdf", mime="application/pdf", use_container_width=True)

    # Keep JSON export the same (as requested)
    with right:
        structured_json = tara_markdown_to_structured_json(
            st.session_state["tara_report"],
            st.session_state.get("last_ecu", selected_ecu)
        )

        json_bytes = json.dumps(
            structured_json,
            indent=2,
            ensure_ascii=False
        ).encode("utf-8")

        st.download_button(
            "Export to JSON",
            data=json_bytes,
            file_name=f"TARA_Report_{sanitize_filename(selected_ecu)}.json",
            mime="application/json",
            use_container_width=True
        )


# ------------------------------
# Damage Scenario Generator (unchanged)
# ------------------------------
def _extract_json_array(text: str):
    """Ultra-robust JSON array extractor for LLM output."""
    import json, re

    # Clean stray artifacts and invisible chars
    text = text.strip().replace("\u200b", "").replace("\xa0", " ")
    text = text.replace("â€œ", '"').replace("â€", '"').replace("â€˜", "'").replace("â€™", "'")

    # Find the longest bracketed JSON array in text
    match = re.search(r'\[.*\]', text, re.DOTALL)
    if not match:
        st.error("âš ï¸ No JSON array found in LLM output.")
        with st.expander("View Raw LLM Output"):
            st.write(text[:6000])
        return []

    json_candidate = match.group(0).strip()

    # Try raw parse
    try:
        return json.loads(json_candidate)
    except json.JSONDecodeError as e:
        # Attempt recovery if missing closing bracket
        st.warning("âš™ï¸ Attempting to auto-fix JSON formatting issues...")
        cleaned = re.sub(r",\s*([\]}])", r"\1", json_candidate)
        cleaned = re.sub(r"([}\]])\s*([{\[])", r"\1,\2", cleaned)
        if not cleaned.endswith("]"):
            cleaned += "]"
        try:
            return json.loads(cleaned)
        except Exception as e2:
            st.error("âŒ Could not parse LLM output even after cleanup.")
            with st.expander("View Raw JSON Candidate"):
                st.text(json_candidate[:8000])
            st.write("Original Error:", str(e))
            st.write("Secondary Error:", str(e2))
            return []

def generate_damage_scenarios_llm(ecu_name: str):
    """LLM-based Damage Scenario Generator (forces raw JSON, no HTML)."""
    prompt = f"""
You are an **automotive cybersecurity analyst** performing Threat Analysis and Risk Assessment (TARA)
as per ISO/SAE 21434.

Your task is to generate **8â€“10 distinct and realistic damage scenarios** for the ECU named: "{ecu_name}".

Output Format Rules:
1. Respond **only** with valid JSON â€” not markdown, not HTML, not text commentary.
2. The JSON must be a single array `[ ... ]` containing multiple scenario objects.
3. Do not include ```json or any code fences.
4. Each object must contain:
   - "title": short scenario title
   - "description": brief summary
   - "cybersecurity_properties_violated": list of violated properties
   - "SFOP": object with four key-value pairs: Severity, Frequency, Observability, Probability

Here is the exact JSON structure to follow:
[
  {{
    "title": "Spoofed SoC readings via CAN",
    "description": "Spoofed SoC and temperature frames cause unsafe overcharge.",
    "cybersecurity_properties_violated": ["Integrity", "Authenticity/Authorization"],
    "SFOP": {{
       "Severity": [5, "High - fire or injury possible"],
       "Frequency": [2, "Requires access, not common"],
       "Observability": [2, "Subtle; often noticed late"],
       "Probability": [3, "Moderate if CAN access exists"]
    }}
  }}
]

Now, generate only the JSON array â€” no text or formatting outside it.
"""
    try:
        response = basic_rag_pipeline.run({
            "text_embedder": {"text": prompt},
            "prompt_builder": {"question": prompt}
        })
        raw_output = response["llm"]["replies"][0]

        # Clean up common LLM formatting artifacts
        raw_output = re.sub(r"```(?:json)?", "", raw_output, flags=re.IGNORECASE)
        raw_output = raw_output.replace("```", "").strip()

        return _extract_json_array(raw_output)

    except Exception as e:
        st.error(f"Error generating scenarios: {e}")
        return []

# ------------------------------
# UI for Damage Scenarios (unchanged)
# ------------------------------
st.markdown("---")
st.markdown("## Damage Scenario Generator")

st.markdown("""
<style>
div[data-baseweb="select"] > div {
    margin-top: -28px;
    min-height: 42px !important;
}
div.stButton > button {
    height: 42px !important;
}
.scenario-grid {
    display: flex;
    flex-wrap: wrap;
    gap: 16px;
    margin-top: 10px;
}
.scenario-card {
    flex: 1 1 calc(50% - 16px);
    box-sizing: border-box;
    border: 1px solid #ccc;
    border-radius: 10px;
    padding: 15px 20px;
    background-color: #f9f9f9;
}
.scenario-card h3 {
    margin-top: 0;
    font-size: 18px;
    font-weight: 600;
}
.scenario-card p {
    margin: 4px 0;
}
.scenario-card ul {
    margin: 6px 0 0 18px;
}
</style>
""", unsafe_allow_html=True)

col1, col2 = st.columns([0.2, 0.8])
with col1:
    num_scenarios = st.selectbox("", [5, 6, 7, 8, 9, 10], index=3, key="num_scenarios_dropdown")
with col2:
    generate_scenarios_clicked = st.button(f"Generate Damage Scenarios for {selected_ecu}", use_container_width=True)

if generate_scenarios_clicked:
    with st.spinner("Generating damage scenarios... please wait"):
        scenarios = generate_damage_scenarios_llm(selected_ecu)
        if not scenarios:
            st.warning("No scenarios generated. Try again.")
        else:
            st.success(f"Generated {num_scenarios} damage scenarios for {selected_ecu}")

            # Build HTML grid for all scenarios
            grid_html = """
            <div class="scenario-grid">
            """
            for s in scenarios[:num_scenarios]:
                sfop_items = "".join(
                    [f"<li><b>{k} ({v[0]}/5)</b> â€“ {v[1]}</li>" for k, v in s["SFOP"].items()]
                )
                grid_html += f"""
                <div class="scenario-card">
                    <h3>{s['title']}</h3>
                    <p><b>Description:</b> {s['description']}</p>
                    <p><b>Cybersecurity Properties Violated:</b> {', '.join(s['cybersecurity_properties_violated'])}</p>
                    <p><b>SFOP Assessment:</b></p>
                    <ul>{sfop_items}</ul>
                </div>
                """
            grid_html += "</div>"

            # Render as pure HTML (no escaping issues)
            st.components.v1.html(f"""
            <html>
            <head>
            <style>
            .scenario-grid {{
                display: flex;
                flex-wrap: wrap;
                gap: 16px;
                margin-top: 10px;
            }}
            .scenario-card {{
                flex: 1 1 calc(50% - 16px);
                font-family: Segoe UI;
                box-sizing: border-box;
                border: 1px solid #ccc;
                border-radius: 10px;
                padding: 15px 20px;
                background-color: #f9f9f9;
                box-shadow: 0 1px 4px rgba(0,0,0,0.08);
            }}
            .scenario-card h3 {{
                margin-top: 0;
                font-size: 18px;
                font-weight: 700;
                color: #222;
            }}
            .scenario-card p {{
                margin: 4px 0;
                font-size: 14px;
                line-height: 1.4;
            }}
            .scenario-card ul {{
                margin: 6px 0 0 18px;
                font-size: 14px;
            }}
            </style>
            </head>
            <body>{grid_html}</body>
            </html>
            """, height=600 + len(scenarios) * 60, scrolling=True)
````

````python
%%writefile app.py
import streamlit as st
from tara_backend import basic_rag_pipeline
from io import BytesIO
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Preformatted,
    PageBreak,
    Table,
    TableStyle,
)
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib import colors
from reportlab.lib.units import mm
from datetime import datetime
import markdown2
import json
import re
import html as html_lib
import unicodedata
from tara_verification import verify_full_tara_report

# NEW: Imports for integrated RAG chatbot
from haystack import Pipeline, Document
from haystack.document_stores.in_memory import InMemoryDocumentStore
from haystack.components.converters import PyPDFToDocument, TextFileToDocument
from haystack.components.preprocessors import DocumentSplitter
from haystack.components.embedders import (
    SentenceTransformersDocumentEmbedder,
    SentenceTransformersTextEmbedder,
)
from haystack.components.retrievers import InMemoryEmbeddingRetriever
from haystack.components.builders import PromptBuilder
from haystack.components.writers import DocumentWriter
import os
import tempfile
from haystack_integrations.components.generators.google_ai import GoogleAIGeminiGenerator

# API key must be available as env var (same as original chatbot)
if "GOOGLE_API_KEY" not in os.environ:
    raise RuntimeError("GOOGLE_API_KEY not set")

# SAME generator as original chatbot
generator = GoogleAIGeminiGenerator(model="gemini-2.5-flash-lite")

# ------------------------------
# ECU Dictionary (unchanged)
# ------------------------------
ECUS = {
    "abs": {"name": "ABS (Anti-lock Braking System)", "type": "ECU"},
    "bms": {"name": "BMS (Battery Management System)", "type": "ECU"},
    "tcu": {"name": "TCU (Telematics Control Unit)", "type": "ECU"},
    "telematics": {"name": "Telematics System", "type": "System"},
    "ecs": {"name": "ECU - General (Electronic Control Unit)", "type": "ECU"},
    "ecm": {"name": "ECM / Powertrain (Engine Control Module)", "type": "ECU"},
    "adas": {"name": "ADAS ECU", "type": "ECU"},
    "infotainment": {"name": "Infotainment Head Unit", "type": "ECU"},
    "gateway": {"name": "Gateway / Domain Controller", "type": "ECU"},
    "eps": {"name": "EPS (Electric Power Steering)", "type": "ECU"},
    "airbag": {"name": "Airbag / SRS ECU", "type": "ECU"},
    "bcm": {"name": "BCM (Body Control Module)", "type": "ECU"},
    "charger": {"name": "EV Charger Interface / Onboard Charger", "type": "Interface"},
    "motor_controller": {"name": "Motor Controller / Inverter", "type": "ECU"},
    "cluster": {"name": "Instrument Cluster", "type": "ECU"},
    "door_control": {"name": "Door Control Module", "type": "ECU"},
    "keyless_entry": {"name": "Keyless Entry / Smart Key System", "type": "Interface"},
    "hvac": {"name": "HVAC Controller", "type": "ECU"},
    "ota_server": {"name": "OTA Server / Update System", "type": "System"},
    "diagnostic": {"name": "Diagnostic Interface (OBD-II / JTAG / UART)", "type": "Interface"},
    "v2x": {"name": "V2X Communication Module", "type": "Interface"},
    "gps": {"name": "GNSS / GPS Receiver", "type": "Interface"},
    "lidar": {"name": "Lidar Sensor System", "type": "System"},
    "radar": {"name": "Radar Sensor System", "type": "System"},
    "camera": {"name": "Camera Sensor System", "type": "System"},
    "tpms": {"name": "TPMS (Tire Pressure Monitoring System)", "type": "System"},
    "obd": {"name": "OBD-II Diagnostic Port", "type": "Interface"},
    "bluetooth": {"name": "Bluetooth Interface", "type": "Interface"},
    "wifi": {"name": "Automotive Wi-Fi Module", "type": "Interface"},
    "can_bus": {"name": "CAN Bus Network", "type": "Interface"},
    "lin_bus": {"name": "LIN Bus Network", "type": "Interface"},
    "flexray": {"name": "FlexRay Network", "type": "Interface"},
    "zonal": {"name": "Zonal ECU", "type": "ECU"},
    "sensor_fusion": {"name": "Sensor Fusion Module", "type": "ECU"},
    "brake_control": {"name": "Brake Control System", "type": "System"},
    "throttle": {"name": "Throttle Control System", "type": "System"},
    "steering": {"name": "Steering ECU", "type": "ECU"},
    "charging_port": {"name": "Automotive Charging Port", "type": "Interface"},
    "mobile_app": {"name": "Vehicle Mobile App", "type": "System"},
    "voice_assistant": {"name": "Automotive Voice Assistant Integration", "type": "System"},
    "rfid_entry": {"name": "RFID Entry System", "type": "Interface"},
    "immobilizer": {"name": "Immobilizer System", "type": "Interface"},
    "ivi": {"name": "IVI (In-Vehicle Infotainment) System", "type": "System"},
    "digital_cockpit": {"name": "Digital Cockpit", "type": "System"},
    "usb": {"name": "USB Interface", "type": "Interface"},
    "hdmi": {"name": "HDMI Port (Automotive IVI)", "type": "Interface"},
    "media_interface": {"name": "Media Player Interface", "type": "System"},
    "e_call": {"name": "E-call System", "type": "System"},
    "telematics_cloud": {"name": "Telematics Cloud API", "type": "System"},
    "vehicle_config": {"name": "Vehicle Configuration API", "type": "System"},
}

# ------------------------------
# Streamlit UI Setup (unchanged)
# ------------------------------
st.set_page_config(page_title="TARA Report Generator", layout="wide")
st.title("ISO/SAE 21434 - TARA Report Generator")
st.markdown("Generate structured Threat Analysis and Risk Assessment (TARA) reports for ECUs using Haystack + Gemini RAG pipeline.")

# ECU Selection (unchanged)
ecu_display_names = [f"{v['name']} [{v['type']}]" for v in ECUS.values()]
ecu_keys = list(ECUS.keys())
selected_display = st.selectbox("Select ECU:", ecu_display_names)
selected_ecu = ECUS[ecu_keys[ecu_display_names.index(selected_display)]]["name"]

if "tara_report" not in st.session_state:
    st.session_state["tara_report"] = ""
    st.session_state["last_ecu"] = ""

# NEW: Session state for chatbot
if "doc_store" not in st.session_state:
    st.session_state.doc_store = InMemoryDocumentStore()
    st.session_state.indexed = False
    st.session_state.chat_history = []
  # ------------------------------
# BASE KNOWLEDGE INGESTION (RUN ONCE)
# ------------------------------
if "base_docs_loaded" not in st.session_state:
    base_files = [
        "mitre_docs.json",
        "atm_docs.json",
        "capec_docs.json",
        "cwe_docs.json",
    ]

    base_documents = []

    for file_name in base_files:
        if os.path.exists(file_name):
            with open(file_name, "r", encoding="utf-8") as f:
                content = json.load(f)
                base_documents.append(
                    Document(
                        content=json.dumps(content, indent=2),
                        meta={"source": file_name},
                    )
                )
        else:
            st.warning(f"âš ï¸ Missing {file_name}")

    if base_documents:
        indexing_pipeline = Pipeline()
        indexing_pipeline.add_component(
            "splitter",
            DocumentSplitter(split_by="word", split_length=250, split_overlap=50),
        )
        indexing_pipeline.add_component(
            "embedder",
            SentenceTransformersDocumentEmbedder(
                model="sentence-transformers/all-MiniLM-L6-v2"
            ),
        )
        indexing_pipeline.add_component(
            "writer", DocumentWriter(st.session_state.doc_store)
        )

        indexing_pipeline.connect("splitter.documents", "embedder.documents")
        indexing_pipeline.connect("embedder.documents", "writer.documents")

        indexing_pipeline.run({"splitter": {"documents": base_documents}})

        st.session_state.indexed = True

    st.session_state.base_docs_loaded = True


# ------------------------------
# Generate TARA Report (unchanged)
# ------------------------------
generate_clicked = st.button("Generate Report", use_container_width=True)
if generate_clicked:
    with st.spinner("Generating report... please wait"):
        try:
            response = basic_rag_pipeline.run(
                {"text_embedder": {"text": selected_ecu}, "prompt_builder": {"question": selected_ecu}}
            )
            report = response["llm"]["replies"][0]
            st.session_state["tara_report"] = report
            st.session_state["last_ecu"] = selected_ecu

            # NEW: Reset chatbot context for new report
            st.session_state.doc_store = InMemoryDocumentStore()
            st.session_state.indexed = False
            st.session_state.chat_history = []

            # NEW: Automatically index the generated report
            report_doc = Document(content=report, meta={"source": "generated_report", "ecu": selected_ecu})
            indexing_pipeline = Pipeline()
            indexing_pipeline.add_component(
                "splitter", DocumentSplitter(split_by="word", split_length=250, split_overlap=50)
            )
            indexing_pipeline.add_component(
                "embedder", SentenceTransformersDocumentEmbedder(model="sentence-transformers/all-MiniLM-L6-v2")
            )
            indexing_pipeline.add_component("writer", DocumentWriter(st.session_state.doc_store))
            indexing_pipeline.connect("splitter.documents", "embedder.documents")
            indexing_pipeline.connect("embedder.documents", "writer.documents")
            indexing_pipeline.run({"splitter": {"documents": [report_doc]}})
            st.session_state.indexed = True

            st.success("TARA report generated successfully!")
        except Exception as e:
            st.error(f"Error generating report: {e}")

# ------------------------------
# Report Preview (unchanged)
# ------------------------------
if st.session_state["tara_report"]:
    st.markdown("**Report Preview:**")

    st.markdown("""
    <style>
    .tara-report-box {
        background-color: #f9f9f9;
        border: 1px solid #ccc;
        border-radius: 10px;
        padding: 20px 25px;
        margin-top: 10px;
        margin-bottom: 25px;
        box-shadow: 0 1px 4px rgba(0,0,0,0.08);
        font-family: 'Segoe UI';
    }
    .tara-report-box h1, .tara-report-box h2, .tara-report-box h3 {
        color: #222;
        margin-top: 0;
    }
    .tara-report-box p {
        color: #333;
        font-size: 15px;
        line-height: 1.5;
    }
    </style>
    """, unsafe_allow_html=True)

    html_content = markdown2.markdown(
        st.session_state["tara_report"],
        extras=["fenced-code-blocks", "tables", "break-on-newline"]
    )
    html_content = re.sub(r"^<div[^>]*>|</div>\s*$", "", html_content.strip())

    st.markdown(f"<div class='tara-report-box'>{html_content}</div>", unsafe_allow_html=True)
else:
    st.info("No report generated yet. Choose an ECU and click Generate Report.")

# ------------------------------
# Helpers for PDF generation
# ------------------------------
def sanitize_filename(name: str):
    name = unicodedata.normalize("NFKD", name)
    name = re.sub(r"[^\w\s-]", "", name).strip()
    name = re.sub(r"[-\s]+", "_", name)
    return name

# Convert markdown to a list of ReportLab flowables; improved table handling
def markdown_to_story(md_text: str, styles, page_width, left_margin, right_margin):
    """
    Convert markdown to a ReportLab story (list of flowables).
    Improved table handling + HTML sanitization for table cells to avoid mismatched tags.
    """
    story = []

    # Convert Markdown -> HTML
    html = markdown2.markdown(md_text, extras=["fenced-code-blocks", "tables", "strike", "cuddled-lists"])
    html = re.sub(r"^<div[^>]*>", "", html)
    html = re.sub(r"</div>\s*$", "", html).strip()

    # Pattern to capture block-level elements
    pattern = re.compile(
        r'(?s)(<h[1-6].*?>.*?</h[1-6]>)|(<pre.*?>.*?</pre>)|(<table.*?>.*?</table>)|(<ul.*?>.*?</ul>)|(<ol.*?>.*?</ul>)|(<p.*?>.*?</p>)',
        flags=re.IGNORECASE,
    )
    matches = list(pattern.finditer(html))

    # fallback if nothing matched
    if not matches:
        cleaned = re.sub(r'<[^>]+>', '', html).strip()
        story.append(Paragraph(html_lib.unescape(cleaned), styles["Normal"]))
        return story

    def inline_replace(s: str) -> str:
        # convert common markdown-generated tags to ReportLab-friendly ones
        s = s.replace("<strong>", "<b>").replace("</strong>", "</b>")
        s = s.replace("<em>", "<i>").replace("</em>", "</i>")
        s = s.replace("<del>", "<u>").replace("</del>", "</u>")
        # convert inline code to font tag
        s = re.sub(r"<code>(.*?)</code>", r"<font name='Courier'>\\1</font>", s, flags=re.DOTALL)
        s = s.replace("<br>", "<br/>")
        s = html_lib.unescape(s)
        return s

    def sanitize_and_balance_cell_html(c_html: str) -> str:
        """Sanitize HTML inside a table cell and balance allowed inline tags."""
        txt = c_html or ""
        txt = html_lib.unescape(txt)

        # Normalize some tags to simpler forms
        txt = txt.replace("<strong>", "<b>").replace("</strong>", "</b>")
        txt = txt.replace("<em>", "<i>").replace("</em>", "</i>")
        txt = txt.replace("<del>", "<u>").replace("</del>", "</u>")
        # handle code -> font
        txt = re.sub(r"<code[^>]*>(.*?)</code>", r"<font name='Courier'>\\1</font>", txt, flags=re.DOTALL)

        # remove any attributes from b/i/u/br tags (not strictly necessary, just clean)
        txt = re.sub(r'<(\/?)((?:b|i|u|br)|font)(?:\s[^>]*)?>', r'<\\1\\2>', txt, flags=re.IGNORECASE)

        # Allow <font name='Courier'> ... </font> (we keep font tags as-is)
        # Remove all tags not in the allowed set {b,i,u,br,font}
        def tag_replacer(match):
            full = match.group(0)
            tagname = match.group(2).lower() if match.group(2) else ""
            # Keep allowed tags, pass through as-is (but strip unknown attrs except font)
            if tagname in ("b", "i", "u", "br"):
                return full  # already cleaned above to no attributes
            if tagname == "font":
                # keep font but strip unsafe attributes (keep name only)
                # extract name attribute if present
                name_match = re.search(r'name\\s*=\\s*[\"\"]([^\"\"]+)[\"\"]', full, flags=re.IGNORECASE)
                if name_match:
                    return f"<font name='{name_match.group(1)}'>"
                # fallback: remove attributes
                return "<font>"
            # anything else -> remove tag entirely
            return ""
        txt = re.sub(r'<\s*(\/?)(\s*[a-zA-Z0-9_:-]+)(?:\s+[^>]*)?>', tag_replacer, txt)

        # Now balance tags for allowed ones to avoid ReportLab parse errors
        def balance_tag(text: str, open_pattern: str, close_tag: str):
            open_count = len(re.findall(open_pattern, text, flags=re.IGNORECASE))
            close_count = text.lower().count(close_tag.lower())
            # remove extra closing tags
            while close_count > open_count:
                text = re.sub(re.escape(close_tag), "", text, count=1, flags=re.IGNORECASE)
                close_count -= 1
            # append missing closing tags
            while open_count > close_count:
                text = text + close_tag
                close_count += 1
            return text

        # Balance <b>, <i>, <u>, <font>
        txt = balance_tag(txt, r'<b(?:\s[^>]*)?>', '</b>')
        txt = balance_tag(txt, r'<i(?:\s[^>]*)?>', '</i>')
        txt = balance_tag(txt, r'<u(?:\s[^>]*)?>', '</u>')
        txt = balance_tag(txt, r'<font(?:\s[^>]*)?>', '</font>')

        # Finally, remove any leftover tags that somehow remain (safety)
        txt = re.sub(r'<(?![/]?[?]?(?:b|i|u|br|font)\\b)[^>]+>', '', txt, flags=re.IGNORECASE)

        # Collapse excessive whitespace
        txt = re.sub(r'\\s+', ' ', txt).strip()
        return txt

    # available width for tables
    usable_width = page_width - (left_margin + right_margin)

    for m in matches:
        block = m.group(0).strip()

        # Headings
        h = re.match(r'^<h([1-6])[^>]*>(.*?)</h\\1>$', block, flags=re.IGNORECASE | re.DOTALL)
        if h:
            lvl = int(h.group(1))
            txt = html_lib.unescape(h.group(2).strip())
            txt = inline_replace(txt)
            if lvl == 1:
                story.append(Paragraph(txt, styles["Title"]))
            elif lvl == 2:
                story.append(Paragraph(txt, styles["Heading2"]))
            elif lvl == 3:
                h3 = ParagraphStyle(name="H3", parent=styles["Heading2"], fontSize=14, leading=16)
                story.append(Paragraph(txt, h3))
            else:
                hsmall = ParagraphStyle(name=f"H{lvl}", parent=styles["Normal"], fontSize=max(9, 14 - lvl))
                story.append(Paragraph(f"<b>{txt}</b>", hsmall))
            story.append(Spacer(1, 6))
            continue

        # Preformatted (code)
        if block.lower().startswith("<pre"):
            code_match = re.search(r'<pre[^>]*>(.*?)</pre>', block, flags=re.IGNORECASE | re.DOTALL)
            code_text = code_match.group(1) if code_match else block
            code_text = re.sub(r'^<code[^>]*>', '', code_text, flags=re.IGNORECASE)
            code_text = re.sub(r'</code>$', '', code_text, flags=re.IGNORECASE)
            code_text = html_lib.unescape(code_text)
            pre_style = ParagraphStyle(name="Code", fontName="Courier", fontSize=9, leading=12)
            story.append(Preformatted(code_text, pre_style))
            story.append(Spacer(1, 6))
            continue

        # Lists
        if block.lower().startswith("<ul") or block.lower().startswith("<ol"):
            items = re.findall(r'<li>(.*?)</li>', block, flags=re.IGNORECASE | re.DOTALL)
            ordered = block.lower().startswith("<ol")
            for i, it in enumerate(items, start=1):
                text = html_lib.unescape(it.strip())
                text = inline_replace(text)
                bullet = f"{i}." if ordered else "â€¢"
                item_style = ParagraphStyle(name="ListItem", parent=styles["Normal"], leftIndent=6)
                story.append(Paragraph(f"{bullet} {text}", item_style))
            story.append(Spacer(1, 6))
            continue

        # Tables: parse rows and render as ReportLab Table (with sanitized cells)
        if block.lower().startswith("<table"):
            rows = re.findall(r'<tr>(.*?)</tr>', block, flags=re.IGNORECASE | re.DOTALL)
            table_data = []
            max_cols = 0
            for row in rows:
                cells = re.findall(r'<t[dh][^>]*>(.*?)</t[dh]>', row, flags=re.IGNORECASE | re.DOTALL)
                clean_cells = []
                for c in cells:
                    # sanitize and balance cell HTML
                    c_text = sanitize_and_balance_cell_html(c)
                    # create Paragraph for the cell (allows inline bold/italic)
                    if c_text.strip() == "":
                        clean_cells.append(Paragraph("", styles["Normal"]))
                    else:
                        clean_cells.append(Paragraph(c_text, styles["Normal"]))
                if len(clean_cells) > max_cols:
                    max_cols = len(clean_cells)
                table_data.append(clean_cells)

            # normalize rows
            for r in table_data:
                if len(r) < max_cols:
                    r.extend([Paragraph("", styles["Normal"])] * (max_cols - len(r)))

            if max_cols == 0:
                continue

            col_width = usable_width / max_cols
            col_widths = [col_width] * max_cols

            tbl = Table(table_data, colWidths=col_widths, repeatRows=1)
            tbl_style = TableStyle([
                ("GRID", (0, 0), (-1, -1), 0.5, colors.HexColor("#cccccc")),
                ("BACKGROUND", (0, 0), (-1, 0), colors.HexColor("#f2f2f2")),
                ("VALIGN", (0, 0), (-1, -1), "TOP"),
                ("LEFTPADDING", (0, 0), (-1, -1), 6),
                ("RIGHTPADDING", (0, 0), (-1, -1), 6),
                ("TOPPADDING", (0, 0), (-1, -1), 4),
                ("BOTTOMPADDING", (0, 0), (-1, -1), 4),
            ])
            tbl.setStyle(tbl_style)
            story.append(tbl)
            story.append(Spacer(1, 8))
            continue

        # Paragraphs
        p_match = re.match(r'^<p[^>]*>(.*?)</p>$', block, flags=re.IGNORECASE | re.DOTALL)
        if p_match:
            inner = p_match.group(1).strip()
            inner = inline_replace(inner)
            story.append(Paragraph(inner, styles["Normal"]))
            story.append(Spacer(1, 6))
            continue

        # Fallback
        cleaned = re.sub(r'<[^>]+>', '', block).strip()
        if cleaned:
            story.append(Paragraph(html_lib.unescape(cleaned), styles["Normal"]))
            story.append(Spacer(1, 6))

    return story

def tara_markdown_to_structured_json(md_text: str, ecu_name: str):
    import re
    from datetime import datetime

    # ---------- ID REGEX ----------
    MITRE_RE = r"\bT\d{4}\b"
    CAPEC_RE = r"\bCAPEC-\d+\b"
    CWE_RE = r"\bCWE-\d+\b"
    ATM_RE = r"\bATM-\d+\b"

    def extract_ids(text, pattern):
        return list(set(re.findall(pattern, text)))

    def clean(v):
        return v.strip().strip("-").strip()

    def to_bool(v):
        v = v.lower()
        return v in ["yes", "true", "impacted"]

    def extract_number(text):
        m = re.findall(r"\d+", text)
        return m[0] if m else ""

    # ---------- BASE STRUCTURE ----------
    data = {
        "metadata": {
            "standard": "ISO/SAE 21434",
            "generated_at": datetime.now().isoformat()
        },
        "ecu": ecu_name,
        "item_definition": {
            "Purpose / Function": "",
            "Operational Environment": "",
            "Interfaces": "",
            "Dependencies": ""
        },
        "assets": [],
        "threats": [],
        "residual_risk": ""
    }

    lines = [l.rstrip() for l in md_text.splitlines() if l.strip()]
    i = 0
    section = None
    current_threat = None

    # ---------- PARSE ----------
    while i < len(lines):
        line = lines[i].strip()

        # ----- SECTION DETECTION -----
        if line.startswith("## 1. Item Definition"):
            section = "item"
            i += 1
            continue

        if line.startswith("## 2. Asset Identification"):
            section = "assets"
            i += 1
            continue

        if line.startswith("## 3. Threat and Risk Assessment"):
            section = "threats"
            i += 1
            continue

        if line.startswith("## 4. Residual Risk"):
            section = "residual"
            i += 1
            continue

        # ----- ITEM DEFINITION -----
        if section == "item" and ":" in line:
            k, v = line.split(":", 1)
            if k.strip() in data["item_definition"]:
                data["item_definition"][k.strip()] = v.strip()

        # ----- ASSETS -----
        if section == "assets" and line.startswith("|") and not line.startswith("| Asset"):
            cols = [c.strip() for c in line.split("|") if c.strip()]
            if len(cols) >= 4 and cols[0] != "------":
                data["assets"].append({
                    "asset": cols[0],
                    "security_property": cols[1],
                    "status": cols[2],
                    "rationale": cols[3]
                })

        # ----- THREAT START -----
        if line.startswith("### Threat ID:"):
            if current_threat:
                data["threats"].append(current_threat)

            tid_title = line.replace("### Threat ID:", "").strip()
            if "â€”" in tid_title:
                tid, title = tid_title.split("â€”", 1)
            else:
                tid, title = tid_title, ""

            current_threat = {
                "threat_id": tid.strip(),
                "title": title.strip(),
                "attributes": {
                    "entry_point_mitre": [],
                    "attack_mechanism_capec": [],
                    "root_cause_cwe": [],
                    "vehicle_relevance_atm": []
                },
                "associated_asset": "",
                "violated_properties": {},
                "impact_rating": "",
                "attack_feasibility_factors": {},
                "feasibility_score": "",
                "risk_score": "",
                "risk_classification": ""
            }

        # ----- THREAT DETAILS -----
        if section == "threats" and current_threat:

            # Attribute rows â†’ ID extraction
            if line.startswith("| Entry Point") or line.startswith("| Attack Mechanism") \
               or line.startswith("| Root Cause") or line.startswith("| Vehicle Relevance"):
                text = lines[i + 1]

                current_threat["attributes"]["entry_point_mitre"] += extract_ids(text, MITRE_RE)
                current_threat["attributes"]["attack_mechanism_capec"] += extract_ids(text, CAPEC_RE)
                current_threat["attributes"]["root_cause_cwe"] += extract_ids(text, CWE_RE)
                current_threat["attributes"]["vehicle_relevance_atm"] += extract_ids(text, ATM_RE)

            # Associated asset
            if line.startswith("| Associated Asset"):
                asset = clean(lines[i + 1])
                if asset and "|" not in asset:
                    current_threat["associated_asset"] = asset

            # CIAAN properties
            if ":" in line and any(p in line for p in [
                "Confidentiality", "Integrity", "Availability",
                "Authenticity", "Authorization",
                "Accountability", "Non-Repudiation"
            ]):
                k, v = line.split(":", 1)
                prop = k.replace("-", "").strip()
                current_threat["violated_properties"][prop] = to_bool(v.strip())

            # Impact rating
            if re.fullmatch(r"[1-5]", line):
                current_threat["impact_rating"] = line

            # Feasibility factors
            if line.startswith("| Required"):
                cols = [c.strip() for c in line.split("|") if c.strip()]
                if len(cols) == 2:
                    current_threat["attack_feasibility_factors"][cols[0]] = extract_number(cols[1])

            # Scores
            if "Feasibility Score" in line:
                current_threat["feasibility_score"] = extract_number(line)

            if "Risk Score" in line:
                current_threat["risk_score"] = extract_number(line)

            if "Risk Classification" in line:
                current_threat["risk_classification"] = line.split("=")[-1].strip().replace("*", "")

        # ----- RESIDUAL RISK -----
        if section == "residual":
            data["residual_risk"] += line + " "

        i += 1

    if current_threat:
        data["threats"].append(current_threat)

    data["residual_risk"] = data["residual_risk"].strip()

    return data

def _extract_json_array(text: str):
    """Ultra-robust JSON array extractor for LLM output."""
    import json, re

    # Clean stray artifacts and invisible chars
    text = text.strip().replace("\u200b", "").replace("\xa0", " ")
    text = text.replace("â€œ", '"').replace("â€", '"').replace("â€˜", "'").replace("â€™", "'")

    # Find the longest bracketed JSON array in text
    match = re.search(r'\[.*\]', text, re.DOTALL)
    if not match:
        st.error("âš ï¸ No JSON array found in LLM output.")
        with st.expander("View Raw LLM Output"):
            st.write(text[:6000])
        return []

    json_candidate = match.group(0).strip()

    # Try raw parse
    try:
        return json.loads(json_candidate)
    except json.JSONDecodeError as e:
        # Attempt recovery if missing closing bracket
        st.warning("âš™ï¸ Attempting to auto-fix JSON formatting issues...")
        cleaned = re.sub(r",\s*([\]}])", r"\1", json_candidate)
        cleaned = re.sub(r"([}\]])\s*([{\[])", r"\1,\2", cleaned)
        if not cleaned.endswith("]"):
            cleaned += "]"
        try:
            return json.loads(cleaned)
        except Exception as e2:
            st.error("âŒ Could not parse LLM output even after cleanup.")
            with st.expander("View Raw JSON Candidate"):
                st.text(json_candidate[:8000])
            st.write("Original Error:", str(e))
            st.write("Secondary Error:", str(e2))
            return []


report_content = st.session_state.get("tara_report", "")

if report_content:
    tara_json = tara_markdown_to_structured_json(report_content, selected_ecu)
    try:
        # Haystack-official way (this is the key fix)
        retriever = basic_rag_pipeline.get_component("retriever")

        verification = verify_full_tara_report(
            tara_json=tara_json,
            retriever=retriever
        )

        # Internal only â€” NOT shown in UI, NOT exported
        with open("tara_verification_log.json", "w") as f:
            json.dump(verification, f, indent=2)

    except Exception as e:
        st.warning(f"Internal verification skipped: {e}")


# ------------------------------
# Export Buttons (improved PDF, landscape)
# ------------------------------
if st.session_state["tara_report"]:
    st.markdown("---")
    left, right = st.columns(2, gap="large")
    with left:
        if st.button("Export to PDF (improved formatting - landscape)", key="export_pdf_landscape"):
            # Landscape A4
            page_size = landscape(A4)
            page_width, page_height = page_size
            left_margin = right_margin = 20 * mm
            top_margin = bottom_margin = 20 * mm

            pdf_buffer = BytesIO()
            doc = SimpleDocTemplate(
                pdf_buffer,
                pagesize=page_size,
                rightMargin=right_margin,
                leftMargin=left_margin,
                topMargin=top_margin,
                bottomMargin=bottom_margin,
            )

            styles = getSampleStyleSheet()
            # Customize styles for headings and normal text
            styles["Title"].fontSize = 20
            styles["Title"].leading = 24
            styles["Heading2"].fontSize = 16
            styles["Heading2"].leading = 20
            styles["Normal"].fontSize = 10
            styles["Normal"].leading = 13

            # Build story
            story = []
            story.append(Paragraph("ISO/SAE 21434 - Threat Analysis and Risk Assessment (TARA) Report", styles["Title"]))
            story.append(Spacer(1, 6))
            story.append(Paragraph(f"<b>ECU:</b> {st.session_state.get('last_ecu', selected_ecu)}", styles["Heading2"]))
            story.append(Spacer(1, 8))

            md_text = st.session_state["tara_report"]
            story += markdown_to_story(md_text, styles, page_width, left_margin, right_margin)

            story.append(Spacer(1, 10))
            story.append(Paragraph(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}", styles["Normal"]))

            doc.build(story)
            pdf_buffer.seek(0)

            safe_name = sanitize_filename(st.session_state.get("last_ecu", selected_ecu))
            st.download_button("Download PDF", data=pdf_buffer, file_name=f"TARA_Report_{safe_name}_landscape.pdf", mime="application/pdf", use_container_width=True)

    # Keep JSON export the same (as requested)
    with right:
        structured_json = tara_markdown_to_structured_json(
            st.session_state["tara_report"],
            st.session_state.get("last_ecu", selected_ecu)
        )

        json_bytes = json.dumps(
            structured_json,
            indent=2,
            ensure_ascii=False
        ).encode("utf-8")

        st.download_button(
            "Export to JSON",
            data=json_bytes,
            file_name=f"TARA_Report_{sanitize_filename(selected_ecu)}.json",
            mime="application/json",
            use_container_width=True
        )


# ------------------------------
# Damage Scenario Generator (unchanged)
# ------------------------------
def generate_damage_scenarios_llm(ecu_name: str):
    """LLM-based Damage Scenario Generator (forces raw JSON, no HTML)."""
    prompt = f"""
You are an **automotive cybersecurity analyst** performing Threat Analysis and Risk Assessment (TARA)
as per ISO/SAE 21434.

Your task is to generate **8â€“10 distinct and realistic damage scenarios** for the ECU named: "{ecu_name}".

Output Format Rules:
1. Respond **only** with valid JSON â€” not markdown, not HTML, not text commentary.
2. The JSON must be a single array `[ ... ]` containing multiple scenario objects.
3. Do not include ```json or any code fences.
4. Each object must contain:
   - "title": short scenario title
   - "description": brief summary
   - "cybersecurity_properties_violated": list of violated properties
   - "SFOP": object with four key-value pairs: Severity, Frequency, Observability, Probability

Here is the exact JSON structure to follow:
[
  {{
    "title": "Spoofed SoC readings via CAN",
    "description": "Spoofed SoC and temperature frames cause unsafe overcharge.",
    "cybersecurity_properties_violated": ["Integrity", "Authenticity/Authorization"],
    "SFOP": {{
       "Severity": [5, "High - fire or injury possible"],
       "Frequency": [2, "Requires access, not common"],
       "Observability": [2, "Subtle; often noticed late"],
       "Probability": [3, "Moderate if CAN access exists"]
    }}
  }}
]

Now, generate only the JSON array â€” no text or formatting outside it.
"""
    try:
        response = basic_rag_pipeline.run({
            "text_embedder": {"text": prompt},
            "prompt_builder": {"question": prompt}
        })
        raw_output = response["llm"]["replies"][0]

        # Clean up common LLM formatting artifacts
        raw_output = re.sub(r"```(?:json)?", "", raw_output, flags=re.IGNORECASE)
        raw_output = raw_output.replace("```", "").strip()

        return _extract_json_array(raw_output)

    except Exception as e:
        st.error(f"Error generating scenarios: {e}")
        return []

# ------------------------------
# UI for Damage Scenarios (unchanged)
# ------------------------------
st.markdown("---")
st.markdown("## Damage Scenario Generator")

st.markdown("""
<style>
div[data-baseweb="select"] > div {
    margin-top: -28px;
    min-height: 42px !important;
}
div.stButton > button {
    height: 42px !important;
}
.scenario-grid {
    display: flex;
    flex-wrap: wrap;
    gap: 16px;
    margin-top: 10px;
}
.scenario-card {
    flex: 1 1 calc(50% - 16px);
    box-sizing: border-box;
    border: 1px solid #ccc;
    border-radius: 10px;
    padding: 15px 20px;
    background-color: #f9f9f9;
}
.scenario-card h3 {
    margin-top: 0;
    font-size: 18px;
    font-weight: 600;
}
.scenario-card p {
    margin: 4px 0;
}
.scenario-card ul {
    margin: 6px 0 0 18px;
}
</style>
""", unsafe_allow_html=True)

col1, col2 = st.columns([0.2, 0.8])
with col1:
    num_scenarios = st.selectbox("", [5, 6, 7, 8, 9, 10], index=3, key="num_scenarios_dropdown")
with col2:
    generate_scenarios_clicked = st.button(f"Generate Damage Scenarios for {selected_ecu}", use_container_width=True)

if generate_scenarios_clicked:
    with st.spinner("Generating damage scenarios... please wait"):
        scenarios = generate_damage_scenarios_llm(selected_ecu)
        if not scenarios:
            st.warning("No scenarios generated. Try again.")
        else:
            st.success(f"Generated {num_scenarios} damage scenarios for {selected_ecu}")

            # Build HTML grid for all scenarios
            grid_html = """
            <div class="scenario-grid">
            """
            for s in scenarios[:num_scenarios]:
                sfop_items = "".join(
                    [f"<li><b>{k} ({v[0]}/5)</b> â€“ {v[1]}</li>" for k, v in s["SFOP"].items()]
                )
                grid_html += f"""
                <div class="scenario-card">
                    <h3>{s['title']}</h3>
                    <p><b>Description:</b> {s['description']}</p>
                    <p><b>Cybersecurity Properties Violated:</b> {', '.join(s['cybersecurity_properties_violated'])}</p>
                    <p><b>SFOP Assessment:</b></p>
                    <ul>{sfop_items}</ul>
                </div>
                """
            grid_html += "</div>"

            # Render as pure HTML (no escaping issues)
            st.components.v1.html(f"""
            <html>
            <head>
            <style>
            .scenario-grid {{
                display: flex;
                flex-wrap: wrap;
                gap: 16px;
                margin-top: 10px;
            }}
            .scenario-card {{
                flex: 1 1 calc(50% - 16px);
                font-family: Segoe UI;
                box-sizing: border-box;
                border: 1px solid #ccc;
                border-radius: 10px;
                padding: 15px 20px;
                background-color: #f9f9f9;
                box-shadow: 0 1px 4px rgba(0,0,0,0.08);
            }}
            .scenario-card h3 {{
                margin-top: 0;
                font-size: 18px;
                font-weight: 700;
                color: #222;
            }}
            .scenario-card p {{
                margin: 4px 0;
                font-size: 14px;
                line-height: 1.4;
            }}
            .scenario-card ul {{
                margin: 6px 0 0 18px;
                font-size: 14px;
            }}
            </style>
            </head>
            <body>{grid_html}</body>
            </html>
            """, height=600 + len(scenarios) * 60, scrolling=True)

# ------------------------------
# NEW: Integrated RAG Chatbot
# ------------------------------
if st.session_state["tara_report"]:  # Only show chatbot after report is generated
    st.markdown("---")
    st.markdown("## Interactive Report Chatbot")
    st.markdown("Ask questions about the generated TARA report or uploaded documents.")

    # File uploader for additional documents (PDF and JSON)
    uploaded_files = st.file_uploader(
        "Upload additional documents (PDF or JSON) to enhance the chatbot's knowledge base",
        type=["pdf", "json"],
        accept_multiple_files=True,
        key="chatbot_uploader"
    )

    # Index additional uploads if provided
    if uploaded_files:
        with st.spinner("Indexing additional documents..."):
            temp_files = []
            json_docs = []
            for f in uploaded_files:
                if f.name.endswith(".pdf"):
                    tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".pdf")
                    tmp.write(f.read())
                    tmp.close()
                    temp_files.append(tmp.name)
                elif f.name.endswith(".json"):
                    # Treat JSON as text document
                    content = json.loads(f.read().decode("utf-8"))
                    json_docs.append(Document(content=json.dumps(content, indent=2), meta={"source": f.name}))

            # Index PDFs
            if temp_files:
                pdf_pipeline = Pipeline()
                pdf_pipeline.add_component("converter", PyPDFToDocument())
                pdf_pipeline.add_component("splitter", DocumentSplitter(split_by="word", split_length=250, split_overlap=50))
                pdf_pipeline.add_component("embedder", SentenceTransformersDocumentEmbedder(model="sentence-transformers/all-MiniLM-L6-v2"))
                pdf_pipeline.add_component("writer", DocumentWriter(st.session_state.doc_store))
                pdf_pipeline.connect("converter.documents", "splitter.documents")
                pdf_pipeline.connect("splitter.documents", "embedder.documents")
                pdf_pipeline.connect("embedder.documents", "writer.documents")
                pdf_pipeline.run({"converter": {"sources": temp_files}})
                for f in temp_files:
                    os.remove(f)

            # Index JSONs directly
            if json_docs:
                json_pipeline = Pipeline()
                json_pipeline.add_component("splitter", DocumentSplitter(split_by="word", split_length=250, split_overlap=50))
                json_pipeline.add_component("embedder", SentenceTransformersDocumentEmbedder(model="sentence-transformers/all-MiniLM-L6-v2"))
                json_pipeline.add_component("writer", DocumentWriter(st.session_state.doc_store))
                json_pipeline.connect("splitter.documents", "embedder.documents")
                json_pipeline.connect("embedder.documents", "writer.documents")
                json_pipeline.run({"splitter": {"documents": json_docs}})

        st.success("Additional documents indexed!")

    # Chat interface
    question = st.text_input("Ask a question about the report or uploaded docs", key="chat_question")
    if question and st.session_state.indexed:
        with st.spinner("Thinking..."):
            # Build chat pipeline
            text_embedder = SentenceTransformersTextEmbedder(model="sentence-transformers/all-MiniLM-L6-v2")
            retriever = InMemoryEmbeddingRetriever(document_store=st.session_state.doc_store, top_k=5)
            prompt = PromptBuilder(
                template="""
You are a helpful technical assistant for automotive cybersecurity TARA reports.

Answer ONLY using the provided context from the generated report and any uploaded documents.
If the answer is not found, say: "Not found in the available context."

Context:
{% for doc in documents %}
{{ doc.content }}
{% endfor %}

Question: {{ question }}
Answer:
"""
            )
            chat_pipeline = Pipeline()
            chat_pipeline.add_component("embedder", text_embedder)
            chat_pipeline.add_component("retriever", retriever)
            chat_pipeline.add_component("prompt_builder", prompt)
            chat_pipeline.add_component("llm", generator)
            chat_pipeline.connect("embedder.embedding", "retriever.query_embedding")
            chat_pipeline.connect("retriever.documents", "prompt_builder.documents")
            chat_pipeline.connect("prompt_builder.prompt", "llm.parts")

            result = chat_pipeline.run(
                {
                    "embedder": {"text": question},
                    "prompt_builder": {"question": question},
                }
            )
            response = result["llm"]["replies"][0]
            st.session_state.chat_history.append({"question": question, "response": response})
    # Display chat history
    if st.session_state.chat_history:
        st.markdown("### Chat History")
        for entry in st.session_state.chat_history:
            st.markdown(f"**Q:** {entry['question']}")
            st.markdown(f"**A:** {entry['response']}")
            st.markdown("---")
````

````python
import json

def convert_bms_to_chatbot_json(data):
    chatbot = {
        "models": [],
        "assets": [],
        "communications": [],
        "attacks": [],
        "cybersecurity": [],
        "damage_scenarios": [],
        "threat_scenarios": [],
        "risk_treatments": []
    }

    # -------------------------------------------------
    # 1ï¸âƒ£ MODELS
    # -------------------------------------------------
    for model in data.get("Models", []):
        chatbot["models"].append({
            "model_name": model.get("name"),
            "created_by": model.get("created_by"),
            "created_at": model.get("Created_at"),
            "status": model.get("status")
        })

    # -------------------------------------------------
    # 2ï¸âƒ£ ASSETS (from Details)
    # -------------------------------------------------
    for asset in data.get("Assets", []):
        for item in asset.get("Details", []):
            chatbot["assets"].append({
                "name": item.get("name"),
                "type": item.get("type"),
                "description": item.get("desc"),
                "security_properties": [
                    p.get("name") for p in item.get("props", [])
                ]
            })

        # -------------------------------------------------
        # 3ï¸âƒ£ COMMUNICATIONS (from template.edges)
        # -------------------------------------------------
        template = asset.get("template", {})
        nodes = template.get("nodes", [])
        edges = template.get("edges", [])

        # Map nodeId â†’ label
        id_to_label = {}
        for node in nodes:
            node_id = node.get("id")
            label = node.get("data", {}).get("label")
            id_to_label[node_id] = label

        for edge in edges:
            chatbot["communications"].append({
                "connection_name": edge.get("data", {}).get("label"),
                "source": id_to_label.get(edge.get("source")),
                "target": id_to_label.get(edge.get("target")),
                "security_properties": edge.get("properties", [])
            })

    # -------------------------------------------------
    # 4ï¸âƒ£ ATTACKS
    # -------------------------------------------------
    for attack in data.get("Attacks", []):
        chatbot["attacks"].append({
            "name": attack.get("name"),
            "description": attack.get("desc") or attack.get("description"),
            "related_asset": attack.get("asset"),
            "severity": attack.get("severity"),
            "likelihood": attack.get("likelihood")
        })

    # -------------------------------------------------
    # 5ï¸âƒ£ CYBERSECURITY
    # -------------------------------------------------
    for cs in data.get("Cybersecurity", []):
        chatbot["cybersecurity"].append({
            "name": cs.get("name"),
            "description": cs.get("desc") or cs.get("description"),
            "related_asset": cs.get("asset")
        })

    # -------------------------------------------------
    # 6ï¸âƒ£ DAMAGE SCENARIOS
    # -------------------------------------------------
    for ds in data.get("Damage_scenarios", []):
        chatbot["damage_scenarios"].append({
            "title": ds.get("title") or ds.get("name"),
            "description": ds.get("description"),
            "impact": ds.get("impact")
        })

    # -------------------------------------------------
    # 7ï¸âƒ£ THREAT SCENARIOS
    # -------------------------------------------------
    for ts in data.get("Threat_scenarios", []):
        chatbot["threat_scenarios"].append({
            "title": ts.get("title") or ts.get("name"),
            "description": ts.get("description"),
            "related_attack": ts.get("attack"),
            "related_asset": ts.get("asset")
        })

    # -------------------------------------------------
    # 8ï¸âƒ£ RISK TREATMENT
    # -------------------------------------------------
    for rt in data.get("Risk_treatment", []):
        chatbot["risk_treatments"].append({
            "risk_id": rt.get("risk_id"),
            "treatment": rt.get("treatment"),
            "status": rt.get("status")
        })

    return chatbot


# -------------------------------------------------
# ğŸ”¥ RUN CONVERSION
# -------------------------------------------------
with open("bms.json", "r", encoding="utf-8") as f:
    raw = json.load(f)

converted = convert_bms_to_chatbot_json(raw)

with open("bms_chatbot_ready.json", "w", encoding="utf-8") as f:
    json.dump(converted, f, indent=2)

print("âœ… Clean chatbot JSON generated: bms_chatbot_ready.json")
````

````python
%%writefile rag_chatbot_app.py

import streamlit as st
from haystack import Pipeline
from haystack.document_stores.in_memory import InMemoryDocumentStore
from haystack.components.converters import PyPDFToDocument
from haystack.components.preprocessors import DocumentSplitter
from haystack.components.embedders import (
    SentenceTransformersDocumentEmbedder,
    SentenceTransformersTextEmbedder,
)
from haystack.components.retrievers import InMemoryEmbeddingRetriever
from haystack.components.builders import PromptBuilder
from haystack.components.writers import DocumentWriter
from haystack import Document
import os
import json
import tempfile
from haystack_integrations.components.generators.google_ai import GoogleAIGeminiGenerator


# -------------------------------
# API Key Check
# -------------------------------
if "GOOGLE_API_KEY" not in os.environ:
    raise RuntimeError("GOOGLE_API_KEY not set")

generator = GoogleAIGeminiGenerator(
    model="gemini-2.5-flash-lite"
)

st.set_page_config(page_title="RAG Chatbot", layout="wide")
st.title("RAG Chatbot")

# -------------------------------
# Session State
# -------------------------------
if "doc_store" not in st.session_state:
    st.session_state.doc_store = InMemoryDocumentStore()
    st.session_state.indexed = False

# -------------------------------
# File Upload (PDF + JSON)
# -------------------------------
uploaded_files = st.file_uploader(
    "Upload PDF or JSON documents",
    type=["pdf", "json"],
    accept_multiple_files=True
)

# -------------------------------
# Index Documents
# -------------------------------
if uploaded_files and not st.session_state.indexed:
    with st.spinner("Indexing documents..."):

        indexing = Pipeline()
        indexing.add_component("splitter",
            DocumentSplitter(split_by="word", split_length=250, split_overlap=50),
        )
        indexing.add_component(
            "embedder",
            SentenceTransformersDocumentEmbedder(
                model="sentence-transformers/all-MiniLM-L6-v2"
            ),
        )
        indexing.add_component("writer", DocumentWriter(st.session_state.doc_store))

        indexing.connect("splitter.documents", "embedder.documents")
        indexing.connect("embedder.documents", "writer.documents")

        documents = []

        for file in uploaded_files:

            # -----------------------
            # Handle PDF
            # -----------------------
            if file.name.endswith(".pdf"):
                tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".pdf")
                tmp.write(file.read())
                tmp.close()

                converter = PyPDFToDocument()
                pdf_docs = converter.run(sources=[tmp.name])["documents"]
                documents.extend(pdf_docs)

                os.remove(tmp.name)

            # -----------------------
            # Handle UTF-8 JSON
            # -----------------------
            elif file.name.endswith(".json"):
                content = file.read().decode("utf-8")
                data = json.loads(content)

                # Case 1: JSON is list of objects
                if isinstance(data, list):
                    for item in data:
                        documents.append(
                            Document(content=json.dumps(item, indent=2))
                        )

                # Case 2: JSON is single object
                elif isinstance(data, dict):
                    documents.append(
                        Document(content=json.dumps(data, indent=2))
                    )

        # Run indexing pipeline
        indexing.run({"splitter": {"documents": documents}})

        st.session_state.indexed = True

    st.success("âœ… Documents indexed. Start chatting.")

# -------------------------------
# Chat Pipeline
# -------------------------------
if st.session_state.indexed:

    text_embedder = SentenceTransformersTextEmbedder(
        model="sentence-transformers/all-MiniLM-L6-v2"
    )

    retriever = InMemoryEmbeddingRetriever(
        document_store=st.session_state.doc_store,
        top_k=5
    )

    prompt = PromptBuilder(
        template="""
You are a strict technical assistant.

IMPORTANT RULES:

1. Answer ONLY using the uploaded document content.
2. If answer is not found, respond exactly:
   Not found in uploaded documents.

3. The output format MUST depend on the Document Type.

Document Type: {{ doc_type }}

FORMAT RULES:

If Document Type is "PDF":
- Return structured technical text.
- Use headings and bullet points.
- Do NOT return JSON.
- Do NOT add extra explanation outside answer.

If Document Type is "JSON":
- Return structured VALID JSON only.
- Do NOT explain.
- Do NOT convert into paragraph format.
- Do NOT add extra text outside JSON.

Context:
{% for doc in documents %}
{{ doc.content }}
{% endfor %}

Question: {{ question }}

Answer:
"""
    )

    chat_pipeline = Pipeline()
    chat_pipeline.add_component("embedder", text_embedder)
    chat_pipeline.add_component("retriever", retriever)
    chat_pipeline.add_component("prompt_builder", prompt)
    chat_pipeline.add_component("llm", generator)

    chat_pipeline.connect("embedder.embedding", "retriever.query_embedding")
    chat_pipeline.connect("retriever.documents", "prompt_builder.documents")
    chat_pipeline.connect("prompt_builder.prompt", "llm.parts")

    question = st.text_input("Ask a question")

    if question:
        with st.spinner("Thinking..."):
            result = chat_pipeline.run(
                {
                    "embedder": {"text": question},
                    "prompt_builder": {"question": question},
                }
            )

        st.markdown("### Retrieved Response")
        st.write(result["llm"]["replies"][0])
````
