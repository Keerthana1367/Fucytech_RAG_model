````python
def ingest_mitre(path, source):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    docs = []
    for obj in data.get("objects", []):
        if obj.get("type") == "attack-pattern":
            name = obj.get("name", "")
            desc = obj.get("description", "")

            if not desc:
                continue

            docs.append(
                Document(
                    content=f"MITRE Technique: {name}\nDescription: {desc}",
                    meta={
                        "source": source,
                        "stix_id": obj.get("id")
                    }
                )
            )
    return docs
````

````python
def ingest_atm(path):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    docs = []
    for obj in data.get("objects", []):
        if obj.get("type") == "attack-pattern":
            name = obj.get("name", "")
            desc = obj.get("description", "")

            if not desc:
                continue

            docs.append(
                Document(
                    content=f"ATM Technique: {name}\nDescription: {desc}",
                    meta={
                        "source": "ATM",
                        "stix_id": obj.get("id")
                    }
                )
            )
    return docs
````

capec

````python
def ingest_capec(xml_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    ns = {"capec": "http://capec.mitre.org/capec-3"}
    docs = []

    for ap in root.findall(".//capec:Attack_Pattern", ns):
        capec_id = ap.get("ID")
        name = ap.findtext("capec:Name", default="", namespaces=ns)
        desc = ap.findtext("capec:Description", default="", namespaces=ns)

        if not desc:
            continue

        docs.append(
            Document(
                content=f"CAPEC-{capec_id}: {name}\nDescription: {desc}",
                meta={
                    "source": "CAPEC",
                    "capec_id": capec_id
                }
            )
        )
    return docs
````

````python
from lxml import etree

def ingest_cwe(xml_path):
    parser = etree.XMLParser(recover=True, huge_tree=True)
    tree = etree.parse(xml_path, parser)
    root = tree.getroot()

    ns = {"cwe": "http://cwe.mitre.org/cwe-7"}
    docs = []

    for w in root.findall(".//cwe:Weakness", namespaces=ns):
        cwe_id = w.get("ID")
        name = w.get("Name")

        desc_elem = w.find("cwe:Description", namespaces=ns)
        desc = desc_elem.text if desc_elem is not None else ""

        if not desc:
            continue

        docs.append(
            Document(
                content=f"CWE-{cwe_id}: {name}\nDescription: {desc}",
                meta={
                    "source": "CWE",
                    "cwe_id": f"CWE-{cwe_id}"
                }
            )
        )

    return docs
````

LOAD AND EMBED JSON DOCS

````python
# ========== 6. LOAD & EMBED DOCUMENTS (with clause-wise JSONs + Annex F) ==========
import json, glob, os
from haystack import Document

# ---------- Paths ----------
CLAUSE_PATH = "./clauses"      # clause-5.json, clause-6.json, ...
ANNEX_PATH  = "annex.json"     # your Annex F JSON file
DATA_PATH   = "data.json"      # ECU/system data

# ---------- Load ECU/System Data ----------
with open(DATA_PATH, "r", encoding="utf-8") as f:
    ecu_data = json.load(f)

# ---------- Load all ISO Clause JSONs ----------
clause_files = sorted(glob.glob(os.path.join(CLAUSE_PATH, "clause-*.json")))
iso_data = []

for file in clause_files:
    with open(file, "r", encoding="utf-8") as f:
        clause_content = json.load(f)

    clause_number = os.path.splitext(os.path.basename(file))[0].split("-")[1]
    iso_data.append({
        "type": "Clause",
        "clause": clause_number,
        "title": clause_content.get("title", f"Clause {clause_number}"),
        "content": clause_content.get("content", json.dumps(clause_content, indent=2))
    })

print(f"📘 Loaded {len(iso_data)} clause JSON files from {CLAUSE_PATH}")

# ---------- Load Annex JSON ----------
annex_data = []
if os.path.exists(ANNEX_PATH):
    with open(ANNEX_PATH, "r", encoding="utf-8") as f:
        annex_json = json.load(f)

    # Handle your Annex F structured JSON
    if isinstance(annex_json, dict) and "annex_id" in annex_json:
        annex_entry = {
            "type": "Annex",
            "annex": annex_json.get("annex_id", "Unknown"),
            "title": annex_json.get("annex_title", "Untitled Annex"),
            "content": ""
        }

        # Flatten and concatenate for embedding
        content_parts = [f"Annex {annex_json.get('annex_id', '')} — {annex_json.get('annex_title', '')}"]
        for section in annex_json.get("sections", []):
            # Section title
            sec_title = f"\n\n### {section.get('section_id', '')} — {section.get('section_title', '')}\n"
            content_parts.append(sec_title)

            # Section content
            for c in section.get("content", []):
                content_parts.append(f"- {c}")

            # Tables
            for t in section.get("tables", []):
                table_header = f"\n\nTable {t.get('table_id', '')}: {t.get('table_title', '')}\n"
                content_parts.append(table_header)

                # Convert rows to readable text
                if "columns" in t and "rows" in t:
                    columns = t["columns"]
                    col_header = " | ".join(columns)
                    content_parts.append(col_header)
                    content_parts.append("-" * len(col_header))
                    for row in t["rows"]:
                        row_line = " | ".join(str(row.get(col, "")) for col in columns)
                        content_parts.append(row_line)

            # Notes
            for note in section.get("notes", []):
                content_parts.append(f"Note: {note}")

        annex_entry["content"] = "\n".join(content_parts)
        annex_data.append(annex_entry)

    elif isinstance(annex_json, list):
        for a in annex_json:
            annex_data.append({
                "type": "Annex",
                "annex": a.get("annex_id", "Unknown"),
                "title": a.get("annex_title", "Untitled Annex"),
                "content": json.dumps(a, indent=2)
            })

    else:
        # Fallback (raw text or simple dict)
        annex_data.append({
            "type": "Annex",
            "annex": "Unknown",
            "title": "Annex (Raw Text)",
            "content": json.dumps(annex_json, indent=2)
        })

    print(f"📗 Loaded {len(annex_data)} annex entries from annex.json")

else:
    print("⚠️  No annex.json found, skipping annex import.")

# ---------- Combine All Data ----------
all_data = list(ecu_data.values()) + iso_data + annex_data
print(f"📄 Total combined documents: {len(all_data)}")

# ---------- Create Haystack Document Objects ----------
docs = []
for item in all_data:
    content = (
        item.get("content")
        or item.get("text")
        or json.dumps(item, indent=2)
    )
    meta = {
        "source": item.get("type", "ECU"),
        "clause": item.get("clause", None),
        "annex": item.get("annex", None),
        "title": item.get("title", "Untitled"),
    }
    docs.append(Document(content=content, meta=meta))

# ---------- Write Raw Documents ----------
document_store.write_documents(docs)
print(f"📄 {len(docs)} documents written to store.")

# ---------- Generate Embeddings ----------
result = doc_embedder.run(documents=docs)
embedded_docs = result["documents"]

# ---------- Save Embedded Documents ----------
document_store.write_documents(embedded_docs, policy="overwrite")
print(f"✅ {len(embedded_docs)} documents embedded and stored successfully.")

# ---------- Summary by Source ----------
summary = {}
for d in embedded_docs:
    src = d.meta.get("source")
    summary[src] = summary.get(src, 0) + 1
print("📊 Document distribution:", summary)
````

````python
%%writefile tara_backend.py
# ==============================================
# TARA BACKEND — STREAMLIT / PRODUCTION VERSION
# ISO + ECU + ANNEX + MITRE + ATM + CAPEC + CWE
# ==============================================

# ========== 1. IMPORTS ==========
import os
import json
import glob
from pathlib import Path

from haystack import Pipeline, Document
from haystack.document_stores.in_memory import InMemoryDocumentStore
from haystack.components.embedders import (
    SentenceTransformersDocumentEmbedder,
    SentenceTransformersTextEmbedder,
)
from haystack.components.retrievers.in_memory import InMemoryEmbeddingRetriever
from haystack.components.builders import PromptBuilder
from haystack_integrations.components.generators.google_ai import GoogleAIGeminiGenerator

print("✅ Imports complete.")

# ========== 2. DOCUMENT STORE & EMBEDDERS ==========
document_store = InMemoryDocumentStore()
print("🧠 Document store initialized.")

doc_embedder = SentenceTransformersDocumentEmbedder(
    model="sentence-transformers/all-MiniLM-L6-v2"
)
text_embedder = SentenceTransformersTextEmbedder(
    model="sentence-transformers/all-MiniLM-L6-v2"
)

doc_embedder.warm_up()
print("✅ Embedders ready.")

retriever = InMemoryEmbeddingRetriever(document_store)
print("🔍 Retriever initialized.")

# ========== 3. STRICT ISO PROMPT ==========
template = """
You are an **automotive cybersecurity analyst** performing a **Threat Analysis and Risk Assessment (TARA)**
strictly according to **ISO/SAE 21434 Clause 15**.

If a property is NOT applicable, explicitly mark it as **Not Applicable (N/A)**.
DO NOT omit any property.

Your task is to generate a **professional, deterministic, and ISO-compliant TARA report**
in **Markdown format ONLY** for the given ECU or system.

⚠️ STRICT RULES (DO NOT VIOLATE):
- Follow the structure EXACTLY as provided
- Do NOT rename sections
- Do NOT add or remove sections
- Do NOT change definitions
- Do NOT invent new scoring models
- Do NOT output explanations, disclaimers, or commentary
- Do NOT vary terminology across runs

Failure to follow these rules makes the output INVALID.

---

### ⚙️ Rules for Reasoning

1. Use **ISO/SAE 21434** to determine structure, compliance requirements, and risk evaluation logic.
2. Use **MITRE ATT&CK (Automotive / ICS / Mobile)** to identify realistic attack vectors and entry points.
3. Use **Automotive Threat Matrix (ATM)** to ensure threats are automotive-specific and ECU-relevant.
4. Use **CAPEC** to describe attack mechanisms and attacker behavior.
5. Use **CWE** to identify underlying software or hardware weaknesses.
6. Always map MITRE / CAPEC / CWE findings back to ISO/SAE 21434 Clause 15 requirements.
7. Do NOT ask the user for missing data — infer using ISO-aligned assumptions.
8. Never invent numbers unless mathematically derived.
9. Output **clean Markdown only** (no code fences, no backticks).

---

## ✅ ISO DEFINITIONS (USE EXACTLY — DO NOT CHANGE)

### Impact Categories (1 = Low → 5 = Severe)
- Safety
- Operational
- Financial
- Privacy
- Regulatory

### Attack Feasibility Factors (1–5 each)
- Required Expertise
- Required Equipment
- Required Knowledge
- Required Access
- Required Time

### Calculations
- **Feasibility Score = (sum of factors ÷ 5)** (rounded to nearest integer)
- **Risk Score = Impact × Feasibility**
- **Risk Classification**
  - 1–5 → Low
  - 6–10 → Medium
  - 11–25 → High

---

## ✅ Cybersecurity Properties (MANDATORY — USE ALL)

For EVERY asset and EVERY threat, evaluate ALL properties below.
If not applicable, explicitly mark **N/A**.

1. Confidentiality
2. Integrity
3. Availability
4. Authenticity
5. Authorization
6. Accountability
7. Non-Repudiation

---

## 📘 ISO CONTEXT (AUTHORITATIVE — DO NOT OVERRIDE)
{% for document in documents %}
- {{ document.content }}
{% endfor %}

System / ECU under analysis: **{{question}}**

---

# ✅ TARA REPORT — {{question}}

---

## 1. Item Definition (Clause 9)
- Purpose / Function
- Operational Environment
- Interfaces
- Dependencies

---

## 2. Asset Identification (Clause 15.3.2)

For EACH asset, produce EXACTLY **7 rows** (one per cybersecurity property).

| Asset | Cybersecurity Property | Status (Impacted / N/A) | Rationale |
|------|------------------------|-------------------------|-----------|

---

## 3. Threat and Risk Assessment (Clauses 15.3.3 – 15.3.7)

Generate **AT LEAST 5 distinct threats**, each with a different attack vector.

---

### Threat ID: TS_<ECU>_00X — <Threat Title>

| Attribute | Value |
|----------|-------|
| Entry Point | Attack vector |
| Associated Asset | Asset |

---

### Cybersecurity Properties Violated
- Confidentiality: Yes / No / N/A
- Integrity: Yes / No / N/A
- Availability: Yes / No / N/A
- Authenticity: Yes / No / N/A
- Authorization: Yes / No / N/A
- Accountability: Yes / No / N/A
- Non-Repudiation: Yes / No / N/A

---

### Damage Scenarios
1. Safety Impact
2. Operational Impact
3. Privacy Impact
4. Financial Impact
5. Regulatory Impact

---

### Impact Rating (1–5)
Provide ONE value based on the most severe damage scenario.

---

### Attack Feasibility Factors (1–5 each)

| Factor | Rating |
|-------|--------|
| Required Expertise | # |
| Required Equipment | # |
| Required Knowledge | # |
| Required Access | # |
| Required Time | # |

► **Feasibility Score = (sum ÷ 5)**
► **Risk Score = Impact × Feasibility**
► **Risk Classification = Low / Medium / High**

---

### Mitigation / Cybersecurity Goals (Clauses 10 & 11)
Map controls explicitly to violated cybersecurity properties.

---

## 4. Residual Risk (Clauses 8 & 11)
Summarize remaining risks and required verification activities.

---
"""

prompt_builder = PromptBuilder(
    template=template,
    required_variables=["documents", "question"]
)

print("🧾 PromptBuilder ready (ISO-locked).")

# ========== 4. GEMINI ==========
if "GOOGLE_API_KEY" not in os.environ:
    raise EnvironmentError("❌ GOOGLE_API_KEY not set")

generator = GoogleAIGeminiGenerator(
    model="gemini-2.5-flash-lite"
)

print("🤖 Gemini initialized.")

# ========== 5. PIPELINE ==========
basic_rag_pipeline = Pipeline()

basic_rag_pipeline.add_component("text_embedder", text_embedder)
basic_rag_pipeline.add_component("retriever", retriever)
basic_rag_pipeline.add_component("prompt_builder", prompt_builder)
basic_rag_pipeline.add_component("llm", generator)

basic_rag_pipeline.connect("text_embedder.embedding", "retriever.query_embedding")
basic_rag_pipeline.connect("retriever", "prompt_builder.documents")
basic_rag_pipeline.connect("prompt_builder", "llm")

print("✅ RAG pipeline built.")

# ========== 6. LOAD DOCUMENTS ==========
CLAUSE_PATH = "./clauses"
ANNEX_PATH = "annex.json"
DATA_PATH = "data.json"

docs = []

# ---------- ECU ----------
with open(DATA_PATH, "r", encoding="utf-8") as f:
    ecu_data = json.load(f)

for name, data in ecu_data.items():
    docs.append(
        Document(
            content=json.dumps(data, indent=2),
            meta={"source": "ECU", "title": name}
        )
    )

print(f"⚙️ ECU docs loaded: {len(ecu_data)}")

# ---------- ISO CLAUSES ----------
for file in glob.glob(os.path.join(CLAUSE_PATH, "clause-*.json")):
    with open(file, "r", encoding="utf-8") as f:
        clause = json.load(f)

    clause_id = Path(file).stem
    docs.append(
        Document(
            content=clause.get("content", json.dumps(clause, indent=2)),
            meta={"source": "ISO_21434", "clause": clause_id}
        )
    )

print("📘 ISO clauses loaded.")

# ---------- ANNEX ----------
if os.path.exists(ANNEX_PATH):
    with open(ANNEX_PATH, "r", encoding="utf-8") as f:
        annex = json.load(f)

    docs.append(
        Document(
            content=json.dumps(annex, indent=2),
            meta={"source": "ANNEX_F"}
        )
    )
    print("📗 Annex F loaded.")

# ---------- HELPER: LOAD PREBUILT THREAT DOCS ----------
def load_docs(path, source):
    loaded = []
    if not os.path.exists(path):
        print(f"⚠️ Missing {path}")
        return loaded

    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    for d in data:
        loaded.append(
            Document(
                content=d["content"],
                meta={**d.get("meta", {}), "source": source}
            )
        )
    return loaded

# ---------- THREAT INTEL ----------
docs += load_docs("mitre_docs.json", "MITRE")
docs += load_docs("atm_docs.json", "ATM")
docs += load_docs("capec_docs.json", "CAPEC")
docs += load_docs("cwe_docs.json", "CWE")

print(f"📄 Total documents before embedding: {len(docs)}")

# ========== 7. EMBED & STORE (ONCE) ==========
embedded_docs = doc_embedder.run(documents=docs)["documents"]
document_store.write_documents(embedded_docs)

print(f"✅ {len(embedded_docs)} documents embedded & stored.")

# ========== 8. EXPORT ==========
__all__ = ["basic_rag_pipeline"]
````

````python
%%writefile app.py
# app.py
import streamlit as st
from tara_backend import basic_rag_pipeline
from io import BytesIO
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Preformatted,
    PageBreak,
    Table,
    TableStyle,
)
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib import colors
from reportlab.lib.units import mm
from datetime import datetime
import markdown2
import json
import re
import html as html_lib
import unicodedata

# ------------------------------
# ECU Dictionary (same as yours)
# ------------------------------
ECUS = {
    "abs": {"name": "ABS (Anti-lock Braking System)", "type": "ECU"},
    "bms": {"name": "BMS (Battery Management System)", "type": "ECU"},
    "tcu": {"name": "TCU (Telematics Control Unit)", "type": "ECU"},
    "telematics": {"name": "Telematics System", "type": "System"},
    "ecs": {"name": "ECU - General (Electronic Control Unit)", "type": "ECU"},
    "ecm": {"name": "ECM / Powertrain (Engine Control Module)", "type": "ECU"},
    "adas": {"name": "ADAS ECU", "type": "ECU"},
    "infotainment": {"name": "Infotainment Head Unit", "type": "ECU"},
    "gateway": {"name": "Gateway / Domain Controller", "type": "ECU"},
    "eps": {"name": "EPS (Electric Power Steering)", "type": "ECU"},
    "airbag": {"name": "Airbag / SRS ECU", "type": "ECU"},
    "bcm": {"name": "BCM (Body Control Module)", "type": "ECU"},
    "charger": {"name": "EV Charger Interface / Onboard Charger", "type": "Interface"},
    "motor_controller": {"name": "Motor Controller / Inverter", "type": "ECU"},
    "cluster": {"name": "Instrument Cluster", "type": "ECU"},
    "door_control": {"name": "Door Control Module", "type": "ECU"},
    "keyless_entry": {"name": "Keyless Entry / Smart Key System", "type": "Interface"},
    "hvac": {"name": "HVAC Controller", "type": "ECU"},
    "ota_server": {"name": "OTA Server / Update System", "type": "System"},
    "diagnostic": {"name": "Diagnostic Interface (OBD-II / JTAG / UART)", "type": "Interface"},
    "v2x": {"name": "V2X Communication Module", "type": "Interface"},
    "gps": {"name": "GNSS / GPS Receiver", "type": "Interface"},
    "lidar": {"name": "Lidar Sensor System", "type": "System"},
    "radar": {"name": "Radar Sensor System", "type": "System"},
    "camera": {"name": "Camera Sensor System", "type": "System"},
    "tpms": {"name": "TPMS (Tire Pressure Monitoring System)", "type": "System"},
    "obd": {"name": "OBD-II Diagnostic Port", "type": "Interface"},
    "bluetooth": {"name": "Bluetooth Interface", "type": "Interface"},
    "wifi": {"name": "Automotive Wi-Fi Module", "type": "Interface"},
    "can_bus": {"name": "CAN Bus Network", "type": "Interface"},
    "lin_bus": {"name": "LIN Bus Network", "type": "Interface"},
    "flexray": {"name": "FlexRay Network", "type": "Interface"},
    "zonal": {"name": "Zonal ECU", "type": "ECU"},
    "sensor_fusion": {"name": "Sensor Fusion Module", "type": "ECU"},
    "brake_control": {"name": "Brake Control System", "type": "System"},
    "throttle": {"name": "Throttle Control System", "type": "System"},
    "steering": {"name": "Steering ECU", "type": "ECU"},
    "charging_port": {"name": "Automotive Charging Port", "type": "Interface"},
    "mobile_app": {"name": "Vehicle Mobile App", "type": "System"},
    "voice_assistant": {"name": "Automotive Voice Assistant Integration", "type": "System"},
    "rfid_entry": {"name": "RFID Entry System", "type": "Interface"},
    "immobilizer": {"name": "Immobilizer System", "type": "Interface"},
    "ivi": {"name": "IVI (In-Vehicle Infotainment) System", "type": "System"},
    "digital_cockpit": {"name": "Digital Cockpit", "type": "System"},
    "usb": {"name": "USB Interface", "type": "Interface"},
    "hdmi": {"name": "HDMI Port (Automotive IVI)", "type": "Interface"},
    "media_interface": {"name": "Media Player Interface", "type": "System"},
    "e_call": {"name": "E-call System", "type": "System"},
    "telematics_cloud": {"name": "Telematics Cloud API", "type": "System"},
    "vehicle_config": {"name": "Vehicle Configuration API", "type": "System"},
}

# ------------------------------
# Streamlit UI Setup
# ------------------------------
st.set_page_config(page_title="TARA Report Generator", layout="wide")
st.title("ISO/SAE 21434 - TARA Report Generator")
st.markdown("Generate structured Threat Analysis and Risk Assessment (TARA) reports for ECUs using Haystack + Gemini RAG pipeline.")

# ECU Selection
ecu_display_names = [f"{v['name']} [{v['type']}]" for v in ECUS.values()]
ecu_keys = list(ECUS.keys())
selected_display = st.selectbox("Select ECU:", ecu_display_names)
selected_ecu = ECUS[ecu_keys[ecu_display_names.index(selected_display)]]["name"]

if "tara_report" not in st.session_state:
    st.session_state["tara_report"] = ""
    st.session_state["last_ecu"] = ""

# ------------------------------
# Generate TARA Report (unchanged)
# ------------------------------
generate_clicked = st.button("Generate Report", use_container_width=True)
if generate_clicked:
    with st.spinner("Generating report... please wait"):
        try:
            response = basic_rag_pipeline.run(
                {"text_embedder": {"text": selected_ecu}, "prompt_builder": {"question": selected_ecu}}
            )
            report = response["llm"]["replies"][0]
            st.session_state["tara_report"] = report
            st.session_state["last_ecu"] = selected_ecu
            st.success("TARA report generated successfully!")
        except Exception as e:
            st.error(f"Error generating report: {e}")

# ------------------------------
# Report Preview (Single Box Style, unchanged)
# ------------------------------
if st.session_state["tara_report"]:
    st.markdown("**Report Preview:**")

    st.markdown("""
    <style>
    .tara-report-box {
        background-color: #f9f9f9;
        border: 1px solid #ccc;
        border-radius: 10px;
        padding: 20px 25px;
        margin-top: 10px;
        margin-bottom: 25px;
        box-shadow: 0 1px 4px rgba(0,0,0,0.08);
        font-family: 'Segoe UI';
    }
    .tara-report-box h1, .tara-report-box h2, .tara-report-box h3 {
        color: #222;
        margin-top: 0;
    }
    .tara-report-box p {
        color: #333;
        font-size: 15px;
        line-height: 1.5;
    }
    </style>
    """, unsafe_allow_html=True)

    html_content = markdown2.markdown(
        st.session_state["tara_report"],
        extras=["fenced-code-blocks", "tables", "break-on-newline"]
    )
    html_content = re.sub(r"^<div[^>]*>|</div>\s*$", "", html_content.strip())

    st.markdown(f"<div class='tara-report-box'>{html_content}</div>", unsafe_allow_html=True)
else:
    st.info("No report generated yet. Choose an ECU and click Generate Report.")

# ------------------------------
# Helpers for PDF generation
# ------------------------------
def sanitize_filename(name: str):
    name = unicodedata.normalize("NFKD", name)
    name = re.sub(r"[^\w\s-]", "", name).strip()
    name = re.sub(r"[-\s]+", "_", name)
    return name

# Convert markdown to a list of ReportLab flowables; improved table handling
# Replace your existing markdown_to_story with this function
def markdown_to_story(md_text: str, styles, page_width, left_margin, right_margin):
    """
    Convert markdown to a ReportLab story (list of flowables).
    Improved table handling + HTML sanitization for table cells to avoid mismatched tags.
    """
    story = []

    # Convert Markdown -> HTML
    html = markdown2.markdown(md_text, extras=["fenced-code-blocks", "tables", "strike", "cuddled-lists"])
    html = re.sub(r"^<div[^>]*>", "", html)
    html = re.sub(r"</div>\s*$", "", html).strip()

    # Pattern to capture block-level elements
    pattern = re.compile(
        r'(?s)(<h[1-6].*?>.*?</h[1-6]>)|(<pre.*?>.*?</pre>)|(<table.*?>.*?</table>)|(<ul.*?>.*?</ul>)|(<ol.*?>.*?</ol>)|(<p.*?>.*?</p>)',
        flags=re.IGNORECASE,
    )
    matches = list(pattern.finditer(html))

    # fallback if nothing matched
    if not matches:
        cleaned = re.sub(r'<[^>]+>', '', html).strip()
        story.append(Paragraph(html_lib.unescape(cleaned), styles["Normal"]))
        return story

    def inline_replace(s: str) -> str:
        # convert common markdown-generated tags to ReportLab-friendly ones
        s = s.replace("<strong>", "<b>").replace("</strong>", "</b>")
        s = s.replace("<em>", "<i>").replace("</em>", "</i>")
        s = s.replace("<del>", "<u>").replace("</del>", "</u>")
        # convert inline code to font tag
        s = re.sub(r"<code>(.*?)</code>", r"<font name='Courier'>\1</font>", s, flags=re.DOTALL)
        s = s.replace("<br>", "<br/>")
        s = html_lib.unescape(s)
        return s

    def sanitize_and_balance_cell_html(c_html: str) -> str:
        """Sanitize HTML inside a table cell and balance allowed inline tags."""
        txt = c_html or ""
        txt = html_lib.unescape(txt)

        # Normalize some tags to simpler forms
        txt = txt.replace("<strong>", "<b>").replace("</strong>", "</b>")
        txt = txt.replace("<em>", "<i>").replace("</em>", "</i>")
        txt = txt.replace("<del>", "<u>").replace("</del>", "</u>")
        # handle code -> font
        txt = re.sub(r"<code[^>]*>(.*?)</code>", r"<font name='Courier'>\1</font>", txt, flags=re.DOTALL)

        # remove any attributes from b/i/u/br tags (not strictly necessary, just clean)
        txt = re.sub(r'<(\/?)(b|i|u|br)\b[^>]*>', r'<\1\2>', txt, flags=re.IGNORECASE)

        # Allow <font name='Courier'> ... </font> (we keep font tags as-is)
        # Remove all tags not in the allowed set {b,i,u,br,font}
        def tag_replacer(match):
            full = match.group(0)
            tagname = match.group(2).lower() if match.group(2) else ""
            # Keep allowed tags, pass through as-is (but strip unknown attrs except font)
            if tagname in ("b", "i", "u", "br"):
                return full  # already cleaned above to no attributes
            if tagname == "font":
                # keep font but strip unsafe attributes (keep name only)
                # extract name attribute if present
                name_match = re.search(r'name\s*=\s*["\']([^"\']+)["\']', full, flags=re.IGNORECASE)
                if name_match:
                    return f"<font name='{name_match.group(1)}'>"
                # fallback: remove attributes
                return "<font>"
            # anything else -> remove tag entirely
            return ""
        txt = re.sub(r'<\s*(\/?)\s*([a-zA-Z0-9_:-]+)(?:\s+[^>]*)?>', tag_replacer, txt)

        # Now balance tags for allowed ones to avoid ReportLab parse errors
        def balance_tag(text: str, open_pattern: str, close_tag: str):
            open_count = len(re.findall(open_pattern, text, flags=re.IGNORECASE))
            close_count = text.lower().count(close_tag.lower())
            # remove extra closing tags
            while close_count > open_count:
                text = re.sub(re.escape(close_tag), "", text, count=1, flags=re.IGNORECASE)
                close_count -= 1
            # append missing closing tags
            while open_count > close_count:
                text = text + close_tag
                close_count += 1
            return text

        # Balance <b>, <i>, <u>, <font>
        txt = balance_tag(txt, r'<b(?:\s[^>]*)?>', '</b>')
        txt = balance_tag(txt, r'<i(?:\s[^>]*)?>', '</i>')
        txt = balance_tag(txt, r'<u(?:\s[^>]*)?>', '</u>')
        txt = balance_tag(txt, r'<font(?:\s[^>]*)?>', '</font>')

        # Finally, remove any leftover tags that somehow remain (safety)
        txt = re.sub(r'<(?!\/?(?:b|i|u|br|font)\b)[^>]+>', '', txt, flags=re.IGNORECASE)

        # Collapse excessive whitespace
        txt = re.sub(r'\s+', ' ', txt).strip()
        return txt

    # available width for tables
    usable_width = page_width - (left_margin + right_margin)

    for m in matches:
        block = m.group(0).strip()

        # Headings
        h = re.match(r'^<h([1-6])[^>]*>(.*?)</h\1>$', block, flags=re.IGNORECASE | re.DOTALL)
        if h:
            lvl = int(h.group(1))
            txt = html_lib.unescape(h.group(2).strip())
            txt = inline_replace(txt)
            if lvl == 1:
                story.append(Paragraph(txt, styles["Title"]))
            elif lvl == 2:
                story.append(Paragraph(txt, styles["Heading2"]))
            elif lvl == 3:
                h3 = ParagraphStyle(name="H3", parent=styles["Heading2"], fontSize=14, leading=16)
                story.append(Paragraph(txt, h3))
            else:
                hsmall = ParagraphStyle(name=f"H{lvl}", parent=styles["Normal"], fontSize=max(9, 14 - lvl))
                story.append(Paragraph(f"<b>{txt}</b>", hsmall))
            story.append(Spacer(1, 6))
            continue

        # Preformatted (code)
        if block.lower().startswith("<pre"):
            code_match = re.search(r'<pre[^>]*>(.*?)</pre>', block, flags=re.IGNORECASE | re.DOTALL)
            code_text = code_match.group(1) if code_match else block
            code_text = re.sub(r'^<code[^>]*>', '', code_text, flags=re.IGNORECASE)
            code_text = re.sub(r'</code>$', '', code_text, flags=re.IGNORECASE)
            code_text = html_lib.unescape(code_text)
            pre_style = ParagraphStyle(name="Code", fontName="Courier", fontSize=9, leading=12)
            story.append(Preformatted(code_text, pre_style))
            story.append(Spacer(1, 6))
            continue

        # Lists
        if block.lower().startswith("<ul") or block.lower().startswith("<ol"):
            items = re.findall(r'<li>(.*?)</li>', block, flags=re.IGNORECASE | re.DOTALL)
            ordered = block.lower().startswith("<ol")
            for i, it in enumerate(items, start=1):
                text = html_lib.unescape(it.strip())
                text = inline_replace(text)
                bullet = f"{i}." if ordered else "•"
                item_style = ParagraphStyle(name="ListItem", parent=styles["Normal"], leftIndent=6)
                story.append(Paragraph(f"{bullet} {text}", item_style))
            story.append(Spacer(1, 6))
            continue

        # Tables: parse rows and render as ReportLab Table (with sanitized cells)
        if block.lower().startswith("<table"):
            rows = re.findall(r'<tr>(.*?)</tr>', block, flags=re.IGNORECASE | re.DOTALL)
            table_data = []
            max_cols = 0
            for row in rows:
                cells = re.findall(r'<t[dh][^>]*>(.*?)</t[dh]>', row, flags=re.IGNORECASE | re.DOTALL)
                clean_cells = []
                for c in cells:
                    # sanitize and balance cell HTML
                    c_text = sanitize_and_balance_cell_html(c)
                    # create Paragraph for the cell (allows inline bold/italic)
                    if c_text.strip() == "":
                        clean_cells.append(Paragraph("", styles["Normal"]))
                    else:
                        clean_cells.append(Paragraph(c_text, styles["Normal"]))
                if len(clean_cells) > max_cols:
                    max_cols = len(clean_cells)
                table_data.append(clean_cells)

            # normalize rows
            for r in table_data:
                if len(r) < max_cols:
                    r.extend([Paragraph("", styles["Normal"])] * (max_cols - len(r)))

            if max_cols == 0:
                continue

            col_width = usable_width / max_cols
            col_widths = [col_width] * max_cols

            tbl = Table(table_data, colWidths=col_widths, repeatRows=1)
            tbl_style = TableStyle([
                ("GRID", (0, 0), (-1, -1), 0.5, colors.HexColor("#cccccc")),
                ("BACKGROUND", (0, 0), (-1, 0), colors.HexColor("#f2f2f2")),
                ("VALIGN", (0, 0), (-1, -1), "TOP"),
                ("LEFTPADDING", (0, 0), (-1, -1), 6),
                ("RIGHTPADDING", (0, 0), (-1, -1), 6),
                ("TOPPADDING", (0, 0), (-1, -1), 4),
                ("BOTTOMPADDING", (0, 0), (-1, -1), 4),
            ])
            tbl.setStyle(tbl_style)
            story.append(tbl)
            story.append(Spacer(1, 8))
            continue

        # Paragraphs
        p_match = re.match(r'^<p[^>]*>(.*?)</p>$', block, flags=re.IGNORECASE | re.DOTALL)
        if p_match:
            inner = p_match.group(1).strip()
            inner = inline_replace(inner)
            story.append(Paragraph(inner, styles["Normal"]))
            story.append(Spacer(1, 6))
            continue

        # Fallback
        cleaned = re.sub(r'<[^>]+>', '', block).strip()
        if cleaned:
            story.append(Paragraph(html_lib.unescape(cleaned), styles["Normal"]))
            story.append(Spacer(1, 6))

    return story

# ------------------------------
# Export Buttons (improved PDF, landscape)
# ------------------------------
if st.session_state["tara_report"]:
    st.markdown("---")
    left, right = st.columns(2, gap="large")
    with left:
        if st.button("Export to PDF (improved formatting - landscape)", key="export_pdf_landscape"):
            # Landscape A4
            page_size = landscape(A4)
            page_width, page_height = page_size
            left_margin = right_margin = 20 * mm
            top_margin = bottom_margin = 20 * mm

            pdf_buffer = BytesIO()
            doc = SimpleDocTemplate(
                pdf_buffer,
                pagesize=page_size,
                rightMargin=right_margin,
                leftMargin=left_margin,
                topMargin=top_margin,
                bottomMargin=bottom_margin,
            )

            styles = getSampleStyleSheet()
            # Customize styles for headings and normal text
            styles["Title"].fontSize = 20
            styles["Title"].leading = 24
            styles["Heading2"].fontSize = 16
            styles["Heading2"].leading = 20
            styles["Normal"].fontSize = 10
            styles["Normal"].leading = 13

            # Build story
            story = []
            story.append(Paragraph("ISO/SAE 21434 - Threat Analysis and Risk Assessment (TARA) Report", styles["Title"]))
            story.append(Spacer(1, 6))
            story.append(Paragraph(f"<b>ECU:</b> {st.session_state.get('last_ecu', selected_ecu)}", styles["Heading2"]))
            story.append(Spacer(1, 8))

            md_text = st.session_state["tara_report"]
            story += markdown_to_story(md_text, styles, page_width, left_margin, right_margin)

            story.append(Spacer(1, 10))
            story.append(Paragraph(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}", styles["Normal"]))

            doc.build(story)
            pdf_buffer.seek(0)

            safe_name = sanitize_filename(st.session_state.get("last_ecu", selected_ecu))
            st.download_button("Download PDF", data=pdf_buffer, file_name=f"TARA_Report_{safe_name}_landscape.pdf", mime="application/pdf", use_container_width=True)

    # Keep JSON export the same (as requested)
    with right:
        report_data = {
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "ecu": st.session_state.get("last_ecu", selected_ecu),
            "tara_report_lines": st.session_state["tara_report"].splitlines(),
        }
        json_bytes = json.dumps(report_data, indent=2, ensure_ascii=False).encode("utf-8")
        st.download_button("Export to JSON", data=json_bytes,
                           file_name=f"TARA_Report_{sanitize_filename(selected_ecu)}.json",
                           mime="application/json", use_container_width=True)

# ------------------------------
# Damage Scenario Generator (unchanged)
# ------------------------------
def _extract_json_array(text: str):
    """Ultra-robust JSON array extractor for LLM output."""
    import json, re

    # Clean stray artifacts and invisible chars
    text = text.strip().replace("\u200b", "").replace("\xa0", " ")
    text = text.replace("“", '"').replace("”", '"').replace("‘", "'").replace("’", "'")

    # Find the longest bracketed JSON array in text
    match = re.search(r'\[.*\]', text, re.DOTALL)
    if not match:
        st.error("⚠️ No JSON array found in LLM output.")
        with st.expander("View Raw LLM Output"):
            st.write(text[:6000])
        return []

    json_candidate = match.group(0).strip()

    # Try raw parse
    try:
        return json.loads(json_candidate)
    except json.JSONDecodeError as e:
        # Attempt recovery if missing closing bracket
        st.warning("⚙️ Attempting to auto-fix JSON formatting issues...")
        cleaned = re.sub(r",\s*([\]}])", r"\1", json_candidate)
        cleaned = re.sub(r"([}\]])\s*([{\[])", r"\1,\2", cleaned)
        if not cleaned.endswith("]"):
            cleaned += "]"
        try:
            return json.loads(cleaned)
        except Exception as e2:
            st.error("❌ Could not parse LLM output even after cleanup.")
            with st.expander("View Raw JSON Candidate"):
                st.text(json_candidate[:8000])
            st.write("Original Error:", str(e))
            st.write("Secondary Error:", str(e2))
            return []

def generate_damage_scenarios_llm(ecu_name: str):
    """LLM-based Damage Scenario Generator (forces raw JSON, no HTML)."""
    prompt = f"""
You are an automotive cybersecurity analyst performing Threat Analysis and Risk Assessment (TARA)
as per ISO/SAE 21434.

Your task is to generate **8–10 distinct and realistic damage scenarios** for the ECU named: "{ecu_name}".

Output Format Rules:
1. Respond **only** with valid JSON — not markdown, not HTML, not text commentary.
2. The JSON must be a single array `[ ... ]` containing multiple scenario objects.
3. Do not include ```json or any code fences.
4. Each object must contain:
   - "title": short scenario title
   - "description": brief summary
   - "cybersecurity_properties_violated": list of violated properties
   - "SFOP": object with four key-value pairs: Severity, Frequency, Observability, Probability

Here is the exact JSON structure to follow:
[
  {{
    "title": "Spoofed SoC readings via CAN",
    "description": "Spoofed SoC and temperature frames cause unsafe overcharge.",
    "cybersecurity_properties_violated": ["Integrity", "Authenticity/Authorization"],
    "SFOP": {{
       "Severity": [5, "High - fire or injury possible"],
       "Frequency": [2, "Requires access, not common"],
       "Observability": [2, "Subtle; often noticed late"],
       "Probability": [3, "Moderate if CAN access exists"]
    }}
  }}
]

Now, generate only the JSON array — no text or formatting outside it.
"""
    try:
        response = basic_rag_pipeline.run({
            "text_embedder": {"text": prompt},
            "prompt_builder": {"question": prompt}
        })
        raw_output = response["llm"]["replies"][0]

        # Clean up common LLM formatting artifacts
        raw_output = re.sub(r"```(?:json)?", "", raw_output, flags=re.IGNORECASE)
        raw_output = raw_output.replace("```", "").strip()

        return _extract_json_array(raw_output)

    except Exception as e:
        st.error(f"Error generating scenarios: {e}")
        return []

# ------------------------------
# UI for Damage Scenarios (unchanged)
# ------------------------------
st.markdown("---")
st.markdown("## Damage Scenario Generator")

st.markdown("""
<style>
div[data-baseweb="select"] > div {
    margin-top: -28px;
    min-height: 42px !important;
}
div.stButton > button {
    height: 42px !important;
}
.scenario-grid {
    display: flex;
    flex-wrap: wrap;
    gap: 16px;
    margin-top: 10px;
}
.scenario-card {
    flex: 1 1 calc(50% - 16px);
    box-sizing: border-box;
    border: 1px solid #ccc;
    border-radius: 10px;
    padding: 15px 20px;
    background-color: #f9f9f9;
}
.scenario-card h3 {
    margin-top: 0;
    font-size: 18px;
    font-weight: 600;
}
.scenario-card p {
    margin: 4px 0;
}
.scenario-card ul {
    margin: 6px 0 0 18px;
}
</style>
""", unsafe_allow_html=True)

col1, col2 = st.columns([0.2, 0.8])
with col1:
    num_scenarios = st.selectbox("", [5, 6, 7, 8, 9, 10], index=3, key="num_scenarios_dropdown")
with col2:
    generate_scenarios_clicked = st.button(f"Generate Damage Scenarios for {selected_ecu}", use_container_width=True)

if generate_scenarios_clicked:
    with st.spinner("Generating damage scenarios... please wait"):
        scenarios = generate_damage_scenarios_llm(selected_ecu)
        if not scenarios:
            st.warning("No scenarios generated. Try again.")
        else:
            st.success(f"Generated {num_scenarios} damage scenarios for {selected_ecu}")

            # Build HTML grid for all scenarios
            grid_html = """
            <div class="scenario-grid">
            """
            for s in scenarios[:num_scenarios]:
                sfop_items = "".join(
                    [f"<li><b>{k} ({v[0]}/5)</b> – {v[1]}</li>" for k, v in s["SFOP"].items()]
                )
                grid_html += f"""
                <div class="scenario-card">
                    <h3>{s['title']}</h3>
                    <p><b>Description:</b> {s['description']}</p>
                    <p><b>Cybersecurity Properties Violated:</b> {', '.join(s['cybersecurity_properties_violated'])}</p>
                    <p><b>SFOP Assessment:</b></p>
                    <ul>{sfop_items}</ul>
                </div>
                """
            grid_html += "</div>"

            # Render as pure HTML (no escaping issues)
            st.components.v1.html(f"""
            <html>
            <head>
            <style>
            .scenario-grid {{
                display: flex;
                flex-wrap: wrap;
                gap: 16px;
                margin-top: 10px;
            }}
            .scenario-card {{
                flex: 1 1 calc(50% - 16px);
                font-family: Segoe UI;
                box-sizing: border-box;
                border: 1px solid #ccc;
                border-radius: 10px;
                padding: 15px 20px;
                background-color: #f9f9f9;
                box-shadow: 0 1px 4px rgba(0,0,0,0.08);
            }}
            .scenario-card h3 {{
                margin-top: 0;
                font-size: 18px;
                font-weight: 700;
                color: #222;
            }}
            .scenario-card p {{
                margin: 4px 0;
                font-size: 14px;
                line-height: 1.4;
            }}
            .scenario-card ul {{
                margin: 6px 0 0 18px;
                font-size: 14px;
            }}
            </style>
            </head>
            <body>{grid_html}</body>
            </html>
            """, height=600 + len(scenarios) * 60, scrolling=True)
````

````python
%%writefile app.py
import streamlit as st
from tara_backend import basic_rag_pipeline
from haystack_integrations.components.generators.google_ai import GoogleAIGeminiGenerator
from io import BytesIO
from datetime import datetime
import markdown2
import json
import re
import html as html_lib
import unicodedata

import os
# ------------------------------
# ECU Dictionary
# ------------------------------
ECUS = {
    "abs": {"name": "ABS (Anti-lock Braking System)", "type": "ECU"},
    "bms": {"name": "BMS (Battery Management System)", "type": "ECU"},
    "tcu": {"name": "TCU (Telematics Control Unit)", "type": "ECU"},
    "telematics": {"name": "Telematics System", "type": "System"},
    "ecs": {"name": "ECU - General (Electronic Control Unit)", "type": "ECU"},
    "ecm": {"name": "ECM / Powertrain (Engine Control Module)", "type": "ECU"},
    "adas": {"name": "ADAS ECU", "type": "ECU"},
    "infotainment": {"name": "Infotainment Head Unit", "type": "ECU"},
    "gateway": {"name": "Gateway / Domain Controller", "type": "ECU"},
    "eps": {"name": "EPS (Electric Power Steering)", "type": "ECU"},
    "airbag": {"name": "Airbag / SRS ECU", "type": "ECU"},
    "bcm": {"name": "BCM (Body Control Module)", "type": "ECU"},
    "charger": {"name": "EV Charger Interface / Onboard Charger", "type": "Interface"},
    "motor_controller": {"name": "Motor Controller / Inverter", "type": "ECU"},
    "cluster": {"name": "Instrument Cluster", "type": "ECU"},
    "door_control": {"name": "Door Control Module", "type": "ECU"},
    "keyless_entry": {"name": "Keyless Entry / Smart Key System", "type": "Interface"},
    "hvac": {"name": "HVAC Controller", "type": "ECU"},
    "ota_server": {"name": "OTA Server / Update System", "type": "System"},
    "diagnostic": {"name": "Diagnostic Interface (OBD-II / JTAG / UART)", "type": "Interface"},
    "v2x": {"name": "V2X Communication Module", "type": "Interface"},
    "gps": {"name": "GNSS / GPS Receiver", "type": "Interface"},
    "lidar": {"name": "Lidar Sensor System", "type": "System"},
    "radar": {"name": "Radar Sensor System", "type": "System"},
    "camera": {"name": "Camera Sensor System", "type": "System"},
    "tpms": {"name": "TPMS (Tire Pressure Monitoring System)", "type": "System"},
    "obd": {"name": "OBD-II Diagnostic Port", "type": "Interface"},
    "bluetooth": {"name": "Bluetooth Interface", "type": "Interface"},
    "wifi": {"name": "Automotive Wi-Fi Module", "type": "Interface"},
    "can_bus": {"name": "CAN Bus Network", "type": "Interface"},
    "lin_bus": {"name": "LIN Bus Network", "type": "Interface"},
    "flexray": {"name": "FlexRay Network", "type": "Interface"},
    "zonal": {"name": "Zonal ECU", "type": "ECU"},
    "sensor_fusion": {"name": "Sensor Fusion Module", "type": "ECU"},
    "brake_control": {"name": "Brake Control System", "type": "System"},
    "throttle": {"name": "Throttle Control System", "type": "System"},
    "steering": {"name": "Steering ECU", "type": "ECU"},
    "charging_port": {"name": "Automotive Charging Port", "type": "Interface"},
    "mobile_app": {"name": "Vehicle Mobile App", "type": "System"},
    "voice_assistant": {"name": "Automotive Voice Assistant Integration", "type": "System"},
    "rfid_entry": {"name": "RFID Entry System", "type": "Interface"},
    "immobilizer": {"name": "Immobilizer System", "type": "Interface"},
    "ivi": {"name": "IVI (In-Vehicle Infotainment) System", "type": "System"},
    "digital_cockpit": {"name": "Digital Cockpit", "type": "System"},
    "usb": {"name": "USB Interface", "type": "Interface"},
    "hdmi": {"name": "HDMI Port (Automotive IVI)", "type": "Interface"},
    "media_interface": {"name": "Media Player Interface", "type": "System"},
    "e_call": {"name": "E-call System", "type": "System"},
    "telematics_cloud": {"name": "Telematics Cloud API", "type": "System"},
    "vehicle_config": {"name": "Vehicle Configuration API", "type": "System"},
}

# ------------------------------
# Streamlit Setup
# ------------------------------
st.set_page_config(page_title="TARA Report Generator", layout="wide")
st.title("ISO/SAE 21434 – TARA Report Generator")

ecu_display = [f"{v['name']} [{v['type']}]" for v in ECUS.values()]
ecu_keys = list(ECUS.keys())
selected_display = st.selectbox("Select ECU:", ecu_display)
selected_ecu = ECUS[ecu_keys[ecu_display.index(selected_display)]]["name"]

if "tara_report" not in st.session_state:
    st.session_state.tara_report = ""
    st.session_state.last_ecu = ""

# ------------------------------
# Generate TARA Report (FIXED)
# ------------------------------
if st.button("Generate TARA Report", use_container_width=True):
    with st.spinner("Generating ISO-compliant TARA report..."):
        try:
            response = basic_rag_pipeline.run(
                {
                    "text_embedder": {"text": selected_ecu},
                    "prompt_builder": {"question": selected_ecu}
                }
            )
            st.session_state.tara_report = response["llm"]["replies"][0]
            st.session_state.last_ecu = selected_ecu
            st.success("TARA report generated successfully.")
        except Exception as e:
            st.error(str(e))

# ------------------------------
# Preview
# ------------------------------
if st.session_state.tara_report:
    st.markdown("## Report Preview")
    html = markdown2.markdown(
        st.session_state.tara_report,
        extras=["tables", "break-on-newline"]
    )
    st.markdown(html, unsafe_allow_html=True)

# ------------------------------
# PDF Export (UNCHANGED)
# ------------------------------
def sanitize_filename(name):
    name = unicodedata.normalize("NFKD", name)
    name = re.sub(r"[^\w\s-]", "", name)
    return re.sub(r"\s+", "_", name)

if st.session_state.tara_report:
    report_data = {
        "timestamp": datetime.now().isoformat(),
        "ecu": st.session_state.last_ecu,
        "report": st.session_state.tara_report,
    }

    st.download_button(
        "Export JSON",
        json.dumps(report_data, indent=2),
        file_name=f"TARA_{sanitize_filename(selected_ecu)}.json",
        mime="application/json",
        use_container_width=True,
    )

# =====================================================
# DAMAGE SCENARIO GENERATOR (LLM ONLY — CORRECT)
# =====================================================
st.markdown("---")
st.markdown("## Damage Scenario Generator")


if "GOOGLE_API_KEY" not in os.environ:
    st.error("GOOGLE_API_KEY not set. Please configure it.")
    st.stop()

gemini = GoogleAIGeminiGenerator(model="gemini-2.5-flash-lite")

def generate_damage_scenarios(ecu):
    prompt = f"""
You are an automotive cybersecurity analyst performing Threat Analysis and Risk Assessment (TARA)
as per ISO/SAE 21434.

Your task is to generate **8–10 distinct and realistic damage scenarios** for the ECU named: "{ecu_name}".

Output Format Rules:
1. Respond **only** with valid JSON — not markdown, not HTML, not text commentary.
2. The JSON must be a single array `[ ... ]` containing multiple scenario objects.
3. Do not include ```json or any code fences.
4. Each object must contain:
   - "title": short scenario title
   - "description": brief summary
   - "cybersecurity_properties_violated": list of violated properties
   - "SFOP": object with four key-value pairs: Severity, Frequency, Observability, Probability

Here is the exact JSON structure to follow:
[
  {{
    "title": "Spoofed SoC readings via CAN",
    "description": "Spoofed SoC and temperature frames cause unsafe overcharge.",
    "cybersecurity_properties_violated": ["Integrity", "Authenticity/Authorization"],
    "SFOP": {{
       "Severity": [5, "High - fire or injury possible"],
       "Frequency": [2, "Requires access, not common"],
       "Observability": [2, "Subtle; often noticed late"],
       "Probability": [3, "Moderate if CAN access exists"]
    }}
  }}
]

Now, generate only the JSON array — no text or formatting outside it.
"""
    result = gemini.run(prompt=prompt)
    raw = result["replies"][0]
    raw = re.sub(r"```.*?```", "", raw, flags=re.DOTALL).strip()
    return json.loads(re.search(r"\[.*\]", raw, re.DOTALL).group(0))

if st.button("Generate Damage Scenarios", use_container_width=True):
    with st.spinner("Generating damage scenarios..."):
        try:
            scenarios = generate_damage_scenarios(selected_ecu)
            for s in scenarios:
                st.markdown(f"### {s['title']}")
                st.write(s["description"])
                st.write("**Violated Properties:**", ", ".join(s["cybersecurity_properties_violated"]))
                st.json(s["SFOP"])
        except Exception as e:
            st.error(str(e))
````
