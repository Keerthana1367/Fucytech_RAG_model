```python
# ==============================
# DATA INGESTION PIPELINE
# ==============================

import json
import xml.etree.ElementTree as ET
from lxml import etree
from haystack import Document
from collections import Counter


# ==============================
# TEXT CHUNKING FUNCTION
# ==============================

def chunk_text(text, chunk_size=400, overlap=50):
    chunks = []
    start = 0

    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start += chunk_size - overlap

    return chunks


# ==============================
# RELATIONSHIP MAP (CAPEC → CWE)
# ==============================

relationships = {
    "CAPEC-66": ["CWE-89"],
    "CAPEC-100": ["CWE-79"],
    "CAPEC-115": ["CWE-119"],
}


# ==============================
# MITRE ATT&CK (Mobile + ICS)
# ==============================

def ingest_mitre(path, source):

    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    docs = []

    for obj in data.get("objects", []):

        if obj.get("type") == "attack-pattern":

            name = obj.get("name", "")
            desc = obj.get("description", "")

            if not desc:
                continue

            text = f"MITRE Technique: {name}\nDescription: {desc}"

            chunks = chunk_text(text)

            for chunk in chunks:

                docs.append(
                    Document(
                        content=chunk,
                        meta={
                            "source": source,
                            "stix_id": obj.get("id"),
                            "technique": name
                        }
                    )
                )

    return docs


mitre_docs = []
mitre_docs += ingest_mitre("mobile-attack (1).json", "MITRE_MOBILE")
mitre_docs += ingest_mitre("ics-attack (1).json", "MITRE_ICS")


# ==============================
# ATM
# ==============================

def ingest_atm(path):

    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    docs = []

    for obj in data.get("objects", []):

        if obj.get("type") == "attack-pattern":

            name = obj.get("name", "")
            desc = obj.get("description", "")

            if not desc:
                continue

            text = f"ATM Technique: {name}\nDescription: {desc}"

            chunks = chunk_text(text)

            for chunk in chunks:

                docs.append(
                    Document(
                        content=chunk,
                        meta={
                            "source": "ATM",
                            "stix_id": obj.get("id"),
                            "technique": name
                        }
                    )
                )

    return docs


atm_docs = ingest_atm("latest atm.json")


# ==============================
# CAPEC
# ==============================

def ingest_capec(xml_path):

    tree = ET.parse(xml_path)
    root = tree.getroot()

    ns = {"capec": "http://capec.mitre.org/capec-3"}

    docs = []

    for ap in root.findall(".//capec:Attack_Pattern", ns):

        capec_id = ap.get("ID")

        name = ap.findtext("capec:Name", default="", namespaces=ns)
        desc = ap.findtext("capec:Description", default="", namespaces=ns)

        if not desc:
            continue

        text = f"CAPEC-{capec_id}: {name}\nDescription: {desc}"

        chunks = chunk_text(text)

        related_cwe = relationships.get(f"CAPEC-{capec_id}", [])

        for chunk in chunks:

            docs.append(
                Document(
                    content=chunk,
                    meta={
                        "source": "CAPEC",
                        "capec_id": f"CAPEC-{capec_id}",
                        "attack_pattern": name,
                        "related_cwe": related_cwe
                    }
                )
            )

    return docs


capec_docs = ingest_capec("capec_v3.9.xml")


# ==============================
# CWE
# ==============================

def ingest_cwe(xml_path):

    parser = etree.XMLParser(recover=True, huge_tree=True)
    tree = etree.parse(xml_path, parser)
    root = tree.getroot()

    ns = {"cwe": "http://cwe.mitre.org/cwe-7"}

    docs = []

    for w in root.findall(".//cwe:Weakness", namespaces=ns):

        cwe_id = w.get("ID")
        name = w.get("Name")

        desc_elem = w.find("cwe:Description", namespaces=ns)

        desc = desc_elem.text if desc_elem is not None else ""

        if not desc:
            continue

        text = f"CWE-{cwe_id}: {name}\nDescription: {desc}"

        chunks = chunk_text(text)

        for chunk in chunks:

            docs.append(
                Document(
                    content=chunk,
                    meta={
                        "source": "CWE",
                        "cwe_id": f"CWE-{cwe_id}",
                        "weakness": name
                    }
                )
            )

    return docs


cwe_docs = ingest_cwe("cwec_v4.19.1.xml")


# ==============================
# MERGE ALL DOCUMENTS
# ==============================

all_docs = []

all_docs += mitre_docs
all_docs += atm_docs
all_docs += capec_docs
all_docs += cwe_docs


# ==============================
# SOURCE COUNT SUMMARY
# ==============================

print("Source distribution:")
print(Counter([d.meta["source"] for d in all_docs]))

print("\nTotal Documents:", len(all_docs))
```

```python
# ============================================================
# CYBERSECURITY DATA INGESTION PIPELINE
# MITRE + ATM + CAPEC + CWE + ISO21434 + Annex + ECU
# ============================================================

import json
import xml.etree.ElementTree as ET
from lxml import etree
from haystack import Document
from collections import Counter
import glob
import os


# ============================================================
# TEXT CHUNKING
# ============================================================

def chunk_text(text, chunk_size=400, overlap=50):

    chunks = []
    start = 0

    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start += chunk_size - overlap

    return chunks


# ============================================================
# RELATIONSHIP MAP
# ============================================================

relationships = {

    "MITRE_TO_CAPEC": {
        "T0855": ["CAPEC-148"],
        "T0842": ["CAPEC-66"]
    },

    "CAPEC_TO_CWE": {
        "CAPEC-66": ["CWE-89"],
        "CAPEC-100": ["CWE-79"],
        "CAPEC-115": ["CWE-119"],
        "CAPEC-148": ["CWE-284"]
    },

    "CWE_TO_ISO": {
        "CWE-284": ["ISO-21434-15"],
        "CWE-79": ["ISO-21434-9"],
        "CWE-89": ["ISO-21434-10"]
    },

    "MITRE_TO_ATM": {
        "T0855": ["ATM-001"],
        "T0842": ["ATM-004"]
    }
}


# ============================================================
# MITRE INGESTION
# ============================================================

def ingest_mitre(path, source):

    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    docs = []

    for obj in data.get("objects", []):

        if obj.get("type") == "attack-pattern":

            technique_id = obj.get("id")
            name = obj.get("name", "")
            desc = obj.get("description", "")

            if not desc:
                continue

            text = f"MITRE Technique: {name}\nDescription: {desc}"

            related_capec = relationships["MITRE_TO_CAPEC"].get(technique_id, [])
            related_atm = relationships["MITRE_TO_ATM"].get(technique_id, [])

            chunks = chunk_text(text)

            for chunk in chunks:

                docs.append(
                    Document(
                        content=chunk,
                        meta={
                            "source": source,
                            "stix_id": technique_id,
                            "technique": name,
                            "related_capec": related_capec,
                            "related_atm": related_atm
                        }
                    )
                )

    return docs


mitre_docs = []
mitre_docs += ingest_mitre("mobile-attack (1).json", "MITRE_MOBILE")
mitre_docs += ingest_mitre("ics-attack (1).json", "MITRE_ICS")


# ============================================================
# ATM INGESTION
# ============================================================

def ingest_atm(path):

    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    docs = []

    for obj in data.get("objects", []):

        if obj.get("type") == "attack-pattern":

            name = obj.get("name", "")
            desc = obj.get("description", "")

            if not desc:
                continue

            text = f"ATM Technique: {name}\nDescription: {desc}"

            chunks = chunk_text(text)

            for chunk in chunks:

                docs.append(
                    Document(
                        content=chunk,
                        meta={
                            "source": "ATM",
                            "stix_id": obj.get("id"),
                            "technique": name
                        }
                    )
                )

    return docs


atm_docs = ingest_atm("latest atm.json")


# ============================================================
# CAPEC INGESTION
# ============================================================

def ingest_capec(xml_path):

    tree = ET.parse(xml_path)
    root = tree.getroot()

    ns = {"capec": "http://capec.mitre.org/capec-3"}

    docs = []

    for ap in root.findall(".//capec:Attack_Pattern", ns):

        capec_id = ap.get("ID")
        name = ap.findtext("capec:Name", default="", namespaces=ns)
        desc = ap.findtext("capec:Description", default="", namespaces=ns)

        if not desc:
            continue

        capec_key = f"CAPEC-{capec_id}"

        text = f"{capec_key}: {name}\nDescription: {desc}"

        related_cwe = relationships["CAPEC_TO_CWE"].get(capec_key, [])

        chunks = chunk_text(text)

        for chunk in chunks:

            docs.append(
                Document(
                    content=chunk,
                    meta={
                        "source": "CAPEC",
                        "capec_id": capec_key,
                        "attack_pattern": name,
                        "related_cwe": related_cwe
                    }
                )
            )

    return docs


capec_docs = ingest_capec("capec_v3.9.xml")


# ============================================================
# CWE INGESTION
# ============================================================

def ingest_cwe(xml_path):

    parser = etree.XMLParser(recover=True, huge_tree=True)
    tree = etree.parse(xml_path, parser)
    root = tree.getroot()

    ns = {"cwe": "http://cwe.mitre.org/cwe-7"}

    docs = []

    for w in root.findall(".//cwe:Weakness", namespaces=ns):

        cwe_id = w.get("ID")
        name = w.get("Name")

        desc_elem = w.find("cwe:Description", namespaces=ns)
        desc = desc_elem.text if desc_elem is not None else ""

        if not desc:
            continue

        cwe_key = f"CWE-{cwe_id}"

        text = f"{cwe_key}: {name}\nDescription: {desc}"

        related_iso = relationships["CWE_TO_ISO"].get(cwe_key, [])

        chunks = chunk_text(text)

        for chunk in chunks:

            docs.append(
                Document(
                    content=chunk,
                    meta={
                        "source": "CWE",
                        "cwe_id": cwe_key,
                        "weakness": name,
                        "related_iso": related_iso
                    }
                )
            )

    return docs


cwe_docs = ingest_cwe("cwec_v4.19.1.xml")


# ============================================================
# ECU DATA INGESTION
# ============================================================

DATA_PATH = "data1.json"

with open(DATA_PATH, "r", encoding="utf-8") as f:
    ecu_data = json.load(f)

ecu_docs = []

for k, v in ecu_data.items():

    text = json.dumps(v, indent=2)

    chunks = chunk_text(text)

    for chunk in chunks:

        ecu_docs.append(
            Document(
                content=chunk,
                meta={
                    "source": "ECU",
                    "title": k
                }
            )
        )


# ============================================================
# ISO 21434 CLAUSES
# ============================================================

CLAUSE_PATH = "./clauses"

clause_files = sorted(glob.glob(os.path.join(CLAUSE_PATH, "clause-*.json")))
iso_docs = []

for file in clause_files:

    with open(file, "r", encoding="utf-8") as f:
        clause_content = json.load(f)

    clause_number = os.path.splitext(os.path.basename(file))[0].split("-")[1]

    text = clause_content.get(
        "content",
        json.dumps(clause_content, indent=2)
    )

    chunks = chunk_text(text)

    for chunk in chunks:

        iso_docs.append(
            Document(
                content=chunk,
                meta={
                    "source": "ISO_21434",
                    "clause": clause_number,
                    "title": clause_content.get("title", f"Clause {clause_number}")
                }
            )
        )


# ============================================================
# ANNEX F
# ============================================================

ANNEX_PATH = "annex.json"
annex_docs = []

if os.path.exists(ANNEX_PATH):

    with open(ANNEX_PATH, "r", encoding="utf-8") as f:
        annex_json = json.load(f)

    content_parts = [
        f"Annex {annex_json.get('annex_id','')}: {annex_json.get('annex_title','')}"
    ]

    for section in annex_json.get("sections", []):

        content_parts.append(
            f"\n\n### {section.get('section_id')} — {section.get('section_title')}"
        )

        for c in section.get("content", []):
            content_parts.append(f"- {c}")

        for note in section.get("notes", []):
            content_parts.append(f"Note: {note}")

    annex_text = "\n".join(content_parts)

    chunks = chunk_text(annex_text)

    for chunk in chunks:

        annex_docs.append(
            Document(
                content=chunk,
                meta={
                    "source": "ANNEX_F",
                    "annex": annex_json.get("annex_id"),
                    "title": annex_json.get("annex_title")
                }
            )
        )


# ============================================================
# MERGE ALL DOCUMENTS
# ============================================================

all_docs = []
all_docs += ecu_docs
all_docs += iso_docs
all_docs += annex_docs
all_docs += mitre_docs
all_docs += atm_docs
all_docs += capec_docs
all_docs += cwe_docs

print("Total documents before embedding:", len(all_docs))


# ============================================================
# EMBEDDING + STORAGE
# ============================================================

result = doc_embedder.run(documents=all_docs)
embedded_docs = result["documents"]

document_store.write_documents(embedded_docs)

print(len(embedded_docs), "documents embedded and stored")


# ============================================================
# SUMMARY
# ============================================================

summary = Counter(d.meta.get("source","Unknown") for d in embedded_docs)

print("\nDocument distribution:")
for k,v in summary.items():
    print(k,":",v)
```
